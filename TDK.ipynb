```python
PLATFORM = "drive"  # drive vagy kaggle, mindkét platformot kipróbáltam

if PLATFORM == "drive":
    try:
        from google.colab import drive
        drive.mount("/content/drive")
    except ImportError:
        print("Drive mount kihagyva")

# Importok

import gc
import json
import os
import random
import warnings
from collections import OrderedDict
from dataclasses import dataclass
from itertools import combinations
from pathlib import Path
from typing import Any, Dict, List, Optional, Tuple

import joblib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
from joblib import Parallel, delayed
from scipy import stats as scipy_stats
from sklearn.isotonic import IsotonicRegression
from sklearn.metrics import (
    balanced_accuracy_score,
    cohen_kappa_score,
    confusion_matrix,
    matthews_corrcoef,
    precision_recall_fscore_support,
)
from sklearn.model_selection import train_test_split
from tensorflow.keras import layers
from tensorflow.keras.applications import EfficientNetB0
from tensorflow.keras.applications.efficientnet import preprocess_input
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from tensorflow.keras.losses import SparseCategoricalCrossentropy
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

warnings.filterwarnings("ignore", category=UserWarning)

AUTOTUNE = tf.data.AUTOTUNE
EPS = 1e-12

# Konfigurációs beállítások

EXPERIMENT_MODE = "FULL"  # Lightweight vagy full, a lightweightet kísérletezés során használtam, gyorsabban fut
RUN_EXPERIMENTS = True
RUN_ANALYSIS = True  # Ha True, analízis/ábrák futnak (RUN_EXPERIMENTS=False esetén lemezről tölti az eredményeket)
RUN_DATASETS = ["cifar10", "shvn", "cifar100"]

GLOBAL_SEEDS = [42, 1337, 2024]
CALIBRATION_METHODS = [
    "uncalibrated",
    "temperature_scaling",
    "static_temperature",
    "vector_scaling",
    "matrix_scaling",
    "isotonic_regression",
]

_PLATFORM_PATHS = {
    "drive": Path("/content/drive/MyDrive/tdk/tdk_final"),
    "kaggle": Path("/kaggle/working/tdk/tdk_final"),
}

BASE_OUTPUT_DIR = _PLATFORM_PATHS[PLATFORM]
BASE_OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
print(f"Platform: {PLATFORM}, kimenet: {BASE_OUTPUT_DIR}")

MODE_CONFIGS = {
    "LIGHTWEIGHT": {
        "datasets": ["cifar10", "svhn", "cifar100"],
        "labeled_ratios": [0.3, 0.5, 0.7],
        "teacher_student_rounds": 3,
        "epochs": 15,
        "batch_size": 128,
        "val_ratio": 0.1,
        "image_size": 224,
        "learning_rate": 1e-4,
        "early_stop_patience": 4,
        "reduce_lr_patience": 2,
        "confidence_threshold": 0.9,
        "entropy_percentile": 30,
        "confidence_threshold_increment": 0.02,
        "entropy_percentile_decrement": 5,
        "pseudo_label_calibration_method": "temperature_scaling",
        "calibration_steps": 250,
        "calibration_lr": 0.01,
        "calibration_l2": 1e-4,
        "calibration_bins": 15,
        "tace_threshold": 0.01,
        "coverage_levels": [0.80, 0.90, 0.95],
        "risk_levels": [0.05, 0.10, 0.20],
        "max_train_samples": 30000,
        "max_test_samples": 10000,
        "strict_dataset_loading": False,
        "n_bootstrap": 1000,
        "significance_alpha": 0.05,
    },
    "FULL": {
        "datasets": ["cifar10", "svhn", "cifar100"],
        "labeled_ratios": [0.1, 0.3, 0.5, 0.7, 0.9],
        "teacher_student_rounds": 3,
        "epochs": 15,
        "batch_size": 128,
        "val_ratio": 0.1,
        "image_size": 224,
        "learning_rate": 1e-4,
        "early_stop_patience": 5,
        "reduce_lr_patience": 2,
        "confidence_threshold": 0.9,
        "entropy_percentile": 30,
        "confidence_threshold_increment": 0.02,
        "entropy_percentile_decrement": 5,
        "pseudo_label_calibration_method": "temperature_scaling",
        "calibration_steps": 350,
        "calibration_lr": 0.01,
        "calibration_l2": 1e-4,
        "calibration_bins": 15,
        "tace_threshold": 0.01,
        "coverage_levels": [0.80, 0.90, 0.95],
        "risk_levels": [0.05, 0.10, 0.20],
        "max_train_samples": 30000,
        "max_test_samples": 10000,
        "strict_dataset_loading": False,
        "n_bootstrap": 2000,
        "significance_alpha": 0.05,
    },
}


def get_experiment_config(mode: str) -> Dict[str, Any]:
    mode = mode.upper()
    if mode not in MODE_CONFIGS:
        raise ValueError(f"Unknown EXPERIMENT_MODE '{mode}'. Use one of {list(MODE_CONFIGS)}.")
    config = dict(MODE_CONFIGS[mode])
    config["mode"] = mode
    config["seeds"] = list(GLOBAL_SEEDS)
    config["datasets"] = list(RUN_DATASETS)
    config["output_dir"] = BASE_OUTPUT_DIR
    config["output_dir"].mkdir(parents=True, exist_ok=True)
    return config


def set_global_seed(seed: int) -> None:
    os.environ["PYTHONHASHSEED"] = str(seed)
    random.seed(seed)
    np.random.seed(seed)
    tf.keras.utils.set_random_seed(seed)
    try:
        tf.config.experimental.enable_op_determinism()
    except Exception:
        pass


def setup_gpu_runtime() -> None:
    gpus = tf.config.list_physical_devices("GPU")
    print("Talált GPU-k:", gpus)
    for gpu in gpus:
        try:
            tf.config.experimental.set_memory_growth(gpu, True)
        except Exception:
            pass
    try:
        tf.keras.mixed_precision.set_global_policy("mixed_float16")
        print("Mixed precision: mixed_float16")
    except Exception as exc:
        print(f"Mixed precision nem érhető el: {exc}")
    try:
        tf.config.optimizer.set_jit(True)
        print("XLA JIT: bekapcsolva")
    except Exception as exc:
        print(f"XLA JIT nem érhető el: {exc}")


setup_gpu_runtime()
ACTIVE_CONFIG = get_experiment_config(EXPERIMENT_MODE)
print(json.dumps({k: v for k, v in ACTIVE_CONFIG.items() if k != "output_dir"}, indent=2))
print("Kimeneti könyvtár:", ACTIVE_CONFIG["output_dir"])

# Adathalmazok

@dataclass
class DatasetBundle:
    name: str
    train_x: np.ndarray
    train_y: np.ndarray
    test_x: np.ndarray
    test_y: np.ndarray
    num_classes: int
    class_names: List[str]
    is_path_dataset: bool = False


def maybe_subsample_stratified(
    x: np.ndarray, y: np.ndarray, max_samples: Optional[int], seed: int,
) -> Tuple[np.ndarray, np.ndarray]:
    if max_samples is None or len(y) <= max_samples:
        return x, y
    idx_keep, _ = train_test_split(
        np.arange(len(y)), train_size=max_samples, stratify=y, random_state=seed,
    )
    return x[idx_keep], y[idx_keep]


def load_cifar10_bundle() -> DatasetBundle:
    (x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar10.load_data()
    y_train = y_train.flatten().astype(np.int32)
    y_test = y_test.flatten().astype(np.int32)
    class_names = [
        "airplane", "automobile", "bird", "cat", "deer",
        "dog", "frog", "horse", "ship", "truck",
    ]
    return DatasetBundle(
        name="cifar10", train_x=x_train.astype(np.uint8), train_y=y_train,
        test_x=x_test.astype(np.uint8), test_y=y_test,
        num_classes=10, class_names=class_names,
    )


def load_cifar100_bundle() -> DatasetBundle:
    (x_train, y_train), (x_test, y_test) = tf.keras.datasets.cifar100.load_data(label_mode="fine")
    y_train = y_train.flatten().astype(np.int32)
    y_test = y_test.flatten().astype(np.int32)
    class_names = [f"class_{i}" for i in range(100)]
    return DatasetBundle(
        name="cifar100", train_x=x_train.astype(np.uint8), train_y=y_train,
        test_x=x_test.astype(np.uint8), test_y=y_test,
        num_classes=100, class_names=class_names,
    )


def _find_file(file_name: str, roots: List[Path]) -> Optional[Path]:
    for root in roots:
        if not root.exists():
            continue
        candidates = list(root.rglob(file_name))
        if candidates:
            return candidates[0]
    return None


def _svhn_search_roots() -> List[Path]:
    """Platformfüggő keresési útvonalak az SVHN .mat fájlokhoz."""
    roots = [Path.cwd()]
    if PLATFORM == "kaggle":
        roots = [Path("/kaggle/input"), Path("/kaggle/working")] + roots
    elif PLATFORM == "drive":
        roots = [Path("/content/drive/MyDrive"), Path("/content")] + roots
    return roots


def load_svhn_bundle() -> DatasetBundle:
    roots = _svhn_search_roots()
    train_mat = _find_file("train_32x32.mat", roots)
    test_mat = _find_file("test_32x32.mat", roots)

    if train_mat is not None and test_mat is not None:
        import scipy.io as sio

        def _load(path: Path) -> Tuple[np.ndarray, np.ndarray]:
            data = sio.loadmat(str(path))
            images = np.transpose(data["X"], (3, 0, 1, 2)).astype(np.uint8)
            labels = data["y"].reshape(-1).astype(np.int32)
            labels = np.where(labels == 10, 0, labels)
            return images, labels

        x_train, y_train = _load(train_mat)
        x_test, y_test = _load(test_mat)
        return DatasetBundle(
            name="svhn", train_x=x_train, train_y=y_train,
            test_x=x_test, test_y=y_test,
            num_classes=10, class_names=[str(i) for i in range(10)],
        )

    print("Az SVHN .mat fájlok nincsenek meg, megpróbálom a tensorflow_datasets betöltést.")
    try:
        import tensorflow_datasets as tfds
    except ImportError as exc:
        raise RuntimeError(
            "SVHN nincs meg helyben, és a tensorflow_datasets sem érhető el."
        ) from exc

    train_ds = tfds.load("svhn_cropped", split="train", as_supervised=True)
    test_ds = tfds.load("svhn_cropped", split="test", as_supervised=True)

    def _collect(ds: tf.data.Dataset) -> Tuple[np.ndarray, np.ndarray]:
        images, labels = [], []
        for image, label in tfds.as_numpy(ds):
            images.append(image)
            labels.append(int(label))
        return np.stack(images).astype(np.uint8), np.asarray(labels, dtype=np.int32)

    x_train, y_train = _collect(train_ds)
    x_test, y_test = _collect(test_ds)
    return DatasetBundle(
        name="svhn", train_x=x_train, train_y=y_train,
        test_x=x_test, test_y=y_test,
        num_classes=10, class_names=[str(i) for i in range(10)],
    )


def _locate_tiny_imagenet_root() -> Path:
    roots = _svhn_search_roots()
    for root in roots:
        if not root.exists():
            continue
        for candidate in root.rglob("tiny-imagenet-200"):
            if (candidate / "train").exists() and (candidate / "val").exists():
                return candidate
    raise FileNotFoundError("Could not find 'tiny-imagenet-200'.")


def load_tiny_imagenet_bundle() -> DatasetBundle:
    tiny_root = _locate_tiny_imagenet_root()
    wnids_path = tiny_root / "wnids.txt"
    if not wnids_path.exists():
        raise FileNotFoundError(f"Missing wnids.txt in {tiny_root}")

    wnids = [line.strip() for line in wnids_path.read_text().splitlines() if line.strip()]
    class_to_idx = {wnid: idx for idx, wnid in enumerate(wnids)}

    train_paths, train_labels = [], []
    for wnid in wnids:
        class_dir = tiny_root / "train" / wnid / "images"
        if not class_dir.exists():
            continue
        image_files = sorted(class_dir.glob("*.JPEG"))
        train_paths.extend([str(p) for p in image_files])
        train_labels.extend([class_to_idx[wnid]] * len(image_files))

    val_annotations = {}
    annotations_file = tiny_root / "val" / "val_annotations.txt"
    if not annotations_file.exists():
        raise FileNotFoundError(f"Missing: {annotations_file}")
    for line in annotations_file.read_text().splitlines():
        parts = line.strip().split("\t")
        if len(parts) >= 2:
            val_annotations[parts[0]] = parts[1]

    val_paths, val_labels = [], []
    for img_path in sorted((tiny_root / "val" / "images").glob("*.JPEG")):
        wnid = val_annotations.get(img_path.name)
        if wnid is None or wnid not in class_to_idx:
            continue
        val_paths.append(str(img_path))
        val_labels.append(class_to_idx[wnid])

    if not train_paths or not val_paths:
        raise RuntimeError("TinyImageNet loader found zero images.")

    return DatasetBundle(
        name="tiny_imagenet",
        train_x=np.asarray(train_paths), train_y=np.asarray(train_labels, dtype=np.int32),
        test_x=np.asarray(val_paths), test_y=np.asarray(val_labels, dtype=np.int32),
        num_classes=len(wnids), class_names=wnids, is_path_dataset=True,
    )


def load_dataset_bundle(dataset_name: str) -> DatasetBundle:
    dataset_name = dataset_name.lower()
    if dataset_name == "cifar10":
        return load_cifar10_bundle()
    if dataset_name == "cifar100":
        return load_cifar100_bundle()
    if dataset_name == "svhn":
        return load_svhn_bundle()
    if dataset_name in {"tiny_imagenet", "tinyimagenet"}:
        return load_tiny_imagenet_bundle()
    raise ValueError(f"Unsupported dataset: {dataset_name}")


# Adatpipeline és augmentáció

def build_data_augmentation(seed: int) -> tf.keras.Sequential:
    return tf.keras.Sequential(
        [
            layers.RandomFlip("horizontal", seed=seed),
            layers.RandomRotation(0.1, seed=seed),
            layers.RandomZoom(0.15, seed=seed),
            layers.RandomTranslation(0.1, 0.1, seed=seed),
        ],
        name=f"augmentation_seed_{seed}",
    )


def preprocess_example(
    image_or_path: tf.Tensor, training: bool, is_path_dataset: bool,
    image_size: int, augmentation: Optional[tf.keras.Sequential],
) -> tf.Tensor:
    if is_path_dataset:
        image_bytes = tf.io.read_file(image_or_path)
        image = tf.io.decode_image(image_bytes, channels=3, expand_animations=False)
        image.set_shape([None, None, 3])
    else:
        image = image_or_path
    image = tf.cast(image, tf.float32)
    image = tf.image.resize(image, (image_size, image_size))
    if training and augmentation is not None:
        image = augmentation(image, training=True)
    image = preprocess_input(image)
    return image


def make_labeled_dataset(
    x: np.ndarray, y: np.ndarray, batch_size: int, is_path_dataset: bool,
    image_size: int, training: bool, seed: int,
    augmentation: Optional[tf.keras.Sequential] = None,
    use_cache: bool = False,
) -> tf.data.Dataset:
    ds = tf.data.Dataset.from_tensor_slices((x, y))
    if training:
        ds = ds.shuffle(buffer_size=min(len(y), 10000), seed=seed, reshuffle_each_iteration=True)
    ds = ds.map(
        lambda image, label: (
            preprocess_example(image, training, is_path_dataset, image_size, augmentation),
            tf.cast(label, tf.int32),
        ),
        num_parallel_calls=AUTOTUNE,
    )
    if use_cache:
        ds = ds.cache()
    ds = ds.batch(batch_size).prefetch(AUTOTUNE)
    return ds


def make_unlabeled_dataset(
    x: np.ndarray, batch_size: int, is_path_dataset: bool, image_size: int,
) -> tf.data.Dataset:
    ds = tf.data.Dataset.from_tensor_slices(x)
    ds = ds.map(
        lambda image: preprocess_example(image, False, is_path_dataset, image_size, None),
        num_parallel_calls=AUTOTUNE,
    )
    ds = ds.batch(batch_size).prefetch(AUTOTUNE)
    return ds


# Modell

def build_classifier(
    num_classes: int, image_size: int, learning_rate: float,
    backbone_trainable: bool = False,
) -> tf.keras.Model:
    backbone = EfficientNetB0(
        weights="imagenet", include_top=False,
        input_shape=(image_size, image_size, 3),
    )
    if not backbone_trainable:
        # Tanármodellnél a backbone fagyasztva marad.
        backbone.trainable = False
    else:
        # Diákmodellnél a felső 30 réteg tanítható, a BN rétegek maradnak fagyasztva.
        backbone.trainable = True
        for _layer in backbone.layers[:-30]:
            _layer.trainable = False
        for _layer in backbone.layers:
            if isinstance(_layer, tf.keras.layers.BatchNormalization):
                _layer.trainable = False

    features = backbone.output
    features = layers.GlobalAveragePooling2D()(features)
    features = layers.Dropout(0.2)(features)
    logits = layers.Dense(num_classes, activation=None, name="logits", dtype="float32")(features)
    model = Model(inputs=backbone.input, outputs=logits)
    model.compile(
        optimizer=Adam(learning_rate=learning_rate),
        loss=SparseCategoricalCrossentropy(from_logits=True),
        metrics=[tf.keras.metrics.SparseCategoricalAccuracy(name="accuracy")],
    )
    return model


# Tanítási segédfüggvények

def train_model(
    model: tf.keras.Model, train_ds: tf.data.Dataset, val_ds: tf.data.Dataset,
    epochs: int, checkpoint_path: Path,
    early_stop_patience: int, reduce_lr_patience: int,
) -> tf.keras.callbacks.History:
    checkpoint_path.parent.mkdir(parents=True, exist_ok=True)
    callbacks = [
        EarlyStopping(monitor="val_loss", patience=early_stop_patience, restore_best_weights=True, verbose=1),
        ReduceLROnPlateau(monitor="val_loss", factor=0.5, patience=reduce_lr_patience, verbose=1),
        ModelCheckpoint(filepath=str(checkpoint_path), monitor="val_loss", save_best_only=True, save_weights_only=True, verbose=0),
    ]
    history = model.fit(train_ds, validation_data=val_ds, epochs=epochs, callbacks=callbacks, verbose=2)
    return history


def predict_logits(model: tf.keras.Model, dataset: tf.data.Dataset, verbose: int = 0) -> np.ndarray:
    return np.asarray(model.predict(dataset, verbose=verbose), dtype=np.float32)


def extract_generalization_gap(history: tf.keras.callbacks.History) -> Dict[str, float]:
    train_acc = float(history.history.get("accuracy", [np.nan])[-1])
    val_acc = float(history.history.get("val_accuracy", [np.nan])[-1])
    train_loss = float(history.history.get("loss", [np.nan])[-1])
    val_loss = float(history.history.get("val_loss", [np.nan])[-1])
    return {
        "train_accuracy": train_acc, "val_accuracy": val_acc,
        "train_loss": train_loss, "val_loss": val_loss,
        "generalization_gap_accuracy": train_acc - val_acc,
        "generalization_gap_loss": train_loss - val_loss,
    }


def save_history(history: tf.keras.callbacks.History, path: Path) -> None:
    path.parent.mkdir(parents=True, exist_ok=True)
    with path.open("w", encoding="utf-8") as handle:
        json.dump(history.history, handle, indent=2)


# Kalibrációs módszerek

def softmax_numpy(logits: np.ndarray) -> np.ndarray:
    logits = np.asarray(logits, dtype=np.float32)
    logits = logits - np.max(logits, axis=1, keepdims=True)
    exp_values = np.exp(logits)
    probs = exp_values / np.sum(exp_values, axis=1, keepdims=True)
    return np.clip(probs, EPS, 1.0)


class BaseCalibrator:
    method_name = "base"

    def fit(self, logits: np.ndarray, y_true: np.ndarray) -> "BaseCalibrator":
        return self

    def predict_proba(self, logits: np.ndarray) -> np.ndarray:
        raise NotImplementedError

    def save(self, path_prefix: Path) -> None:
        path_prefix.parent.mkdir(parents=True, exist_ok=True)
        with (path_prefix.with_suffix(".json")).open("w", encoding="utf-8") as handle:
            json.dump({"method": self.method_name}, handle, indent=2)


class IdentityCalibrator(BaseCalibrator):
    method_name = "uncalibrated"

    def predict_proba(self, logits: np.ndarray) -> np.ndarray:
        return softmax_numpy(logits)


class TemperatureScalingCalibrator(BaseCalibrator):
    method_name = "temperature_scaling"

    def __init__(self, steps: int = 300, learning_rate: float = 0.01):
        self.steps = steps
        self.learning_rate = learning_rate
        self.log_temperature = 0.0

    @property
    def temperature(self) -> float:
        return float(np.exp(self.log_temperature))

    def fit(self, logits: np.ndarray, y_true: np.ndarray) -> "TemperatureScalingCalibrator":
        logits_tf = tf.convert_to_tensor(logits, dtype=tf.float32)
        labels_tf = tf.convert_to_tensor(y_true, dtype=tf.int32)
        log_temperature = tf.Variable(0.0, dtype=tf.float32)
        optimizer = tf.keras.optimizers.Adam(self.learning_rate)
        for _ in range(self.steps):
            with tf.GradientTape() as tape:
                scaled_logits = logits_tf / tf.exp(log_temperature)
                loss = tf.reduce_mean(
                    tf.keras.losses.sparse_categorical_crossentropy(labels_tf, scaled_logits, from_logits=True)
                )
            gradients = tape.gradient(loss, [log_temperature])
            optimizer.apply_gradients(zip(gradients, [log_temperature]))
        self.log_temperature = float(log_temperature.numpy())
        return self

    def predict_proba(self, logits: np.ndarray) -> np.ndarray:
        temperature = max(self.temperature, 1e-6)
        scaled_logits = np.asarray(logits, dtype=np.float32) / temperature
        return softmax_numpy(scaled_logits)

    def save(self, path_prefix: Path) -> None:
        path_prefix.parent.mkdir(parents=True, exist_ok=True)
        payload = {"method": self.method_name, "log_temperature": self.log_temperature, "temperature": self.temperature}
        with (path_prefix.with_suffix(".json")).open("w", encoding="utf-8") as handle:
            json.dump(payload, handle, indent=2)


class StaticTemperatureCalibrator(BaseCalibrator):
    """Statikus hőmérséklet-kalibrátor fix T értékkel."""
    method_name = "static_temperature"

    def __init__(self, temperature: float = 1.5):
        self.fixed_temperature = temperature

    def predict_proba(self, logits: np.ndarray) -> np.ndarray:
        scaled_logits = np.asarray(logits, dtype=np.float32) / max(self.fixed_temperature, 1e-6)
        return softmax_numpy(scaled_logits)

    def save(self, path_prefix: Path) -> None:
        path_prefix.parent.mkdir(parents=True, exist_ok=True)
        payload = {"method": self.method_name, "temperature": self.fixed_temperature}
        with (path_prefix.with_suffix(".json")).open("w", encoding="utf-8") as handle:
            json.dump(payload, handle, indent=2)


class VectorScalingCalibrator(BaseCalibrator):
    method_name = "vector_scaling"

    def __init__(self, num_classes: int, steps: int = 300, learning_rate: float = 0.01, l2: float = 1e-4):
        self.num_classes = num_classes
        self.steps = steps
        self.learning_rate = learning_rate
        self.l2 = l2
        self.scale = np.ones(num_classes, dtype=np.float32)
        self.bias = np.zeros(num_classes, dtype=np.float32)

    def fit(self, logits: np.ndarray, y_true: np.ndarray) -> "VectorScalingCalibrator":
        logits_tf = tf.convert_to_tensor(logits, dtype=tf.float32)
        labels_tf = tf.convert_to_tensor(y_true, dtype=tf.int32)
        log_scale = tf.Variable(tf.zeros([self.num_classes], dtype=tf.float32))
        bias = tf.Variable(tf.zeros([self.num_classes], dtype=tf.float32))
        optimizer = tf.keras.optimizers.Adam(self.learning_rate)
        for _ in range(self.steps):
            with tf.GradientTape() as tape:
                scaled_logits = logits_tf * tf.exp(log_scale) + bias
                nll = tf.reduce_mean(tf.keras.losses.sparse_categorical_crossentropy(labels_tf, scaled_logits, from_logits=True))
                reg = self.l2 * (tf.reduce_sum(tf.square(log_scale)) + tf.reduce_sum(tf.square(bias)))
                loss = nll + reg
            grads = tape.gradient(loss, [log_scale, bias])
            optimizer.apply_gradients(zip(grads, [log_scale, bias]))
        self.scale = np.exp(log_scale.numpy()).astype(np.float32)
        self.bias = bias.numpy().astype(np.float32)
        return self

    def predict_proba(self, logits: np.ndarray) -> np.ndarray:
        logits = np.asarray(logits, dtype=np.float32)
        return softmax_numpy(logits * self.scale + self.bias)

    def save(self, path_prefix: Path) -> None:
        path_prefix.parent.mkdir(parents=True, exist_ok=True)
        np.savez(str(path_prefix.with_suffix(".npz")), scale=self.scale, bias=self.bias)
        payload = {"method": self.method_name, "num_classes": int(self.num_classes)}
        with (path_prefix.with_suffix(".json")).open("w", encoding="utf-8") as handle:
            json.dump(payload, handle, indent=2)


class MatrixScalingCalibrator(BaseCalibrator):
    method_name = "matrix_scaling"

    def __init__(self, num_classes: int, steps: int = 300, learning_rate: float = 0.01, l2: float = 1e-4):
        self.num_classes = num_classes
        self.steps = steps
        self.learning_rate = learning_rate
        self.l2 = l2
        self.weight = np.eye(num_classes, dtype=np.float32)
        self.bias = np.zeros(num_classes, dtype=np.float32)

    def fit(self, logits: np.ndarray, y_true: np.ndarray) -> "MatrixScalingCalibrator":
        logits_tf = tf.convert_to_tensor(logits, dtype=tf.float32)
        labels_tf = tf.convert_to_tensor(y_true, dtype=tf.int32)
        weight = tf.Variable(tf.eye(self.num_classes, dtype=tf.float32))
        bias = tf.Variable(tf.zeros([self.num_classes], dtype=tf.float32))
        optimizer = tf.keras.optimizers.Adam(self.learning_rate)
        eye = tf.eye(self.num_classes, dtype=tf.float32)
        for _ in range(self.steps):
            with tf.GradientTape() as tape:
                scaled_logits = tf.linalg.matmul(logits_tf, weight) + bias
                nll = tf.reduce_mean(tf.keras.losses.sparse_categorical_crossentropy(labels_tf, scaled_logits, from_logits=True))
                reg = self.l2 * (tf.reduce_sum(tf.square(weight - eye)) + tf.reduce_sum(tf.square(bias)))
                loss = nll + reg
            grads = tape.gradient(loss, [weight, bias])
            optimizer.apply_gradients(zip(grads, [weight, bias]))
        self.weight = weight.numpy().astype(np.float32)
        self.bias = bias.numpy().astype(np.float32)
        return self

    def predict_proba(self, logits: np.ndarray) -> np.ndarray:
        logits = np.asarray(logits, dtype=np.float32)
        return softmax_numpy(np.matmul(logits, self.weight) + self.bias)

    def save(self, path_prefix: Path) -> None:
        path_prefix.parent.mkdir(parents=True, exist_ok=True)
        np.savez(str(path_prefix.with_suffix(".npz")), weight=self.weight, bias=self.bias)
        payload = {"method": self.method_name, "num_classes": int(self.num_classes)}
        with (path_prefix.with_suffix(".json")).open("w", encoding="utf-8") as handle:
            json.dump(payload, handle, indent=2)


def _fit_one_isotonic(class_idx, probs, y_true):
    """Egy osztály isotonic regressziója párhuzamosításhoz."""
    binary = (y_true == class_idx).astype(np.int32)
    cp = probs[:, class_idx]
    if np.unique(binary).size < 2:
        return class_idx, None, float(np.mean(binary))
    m = IsotonicRegression(out_of_bounds="clip", y_min=0.0, y_max=1.0)
    m.fit(cp, binary)
    return class_idx, m, float(np.mean(binary))


class FastIsotonicCalibrator(BaseCalibrator):
    """Párhuzamos isotonic regression kalibrátor."""
    method_name = "isotonic_regression"

    def __init__(self, num_classes: int):
        self.num_classes = num_classes
        self.models: List[Optional[IsotonicRegression]] = []
        self.constant_values: List[float] = []

    def fit(self, logits: np.ndarray, y_true: np.ndarray) -> "FastIsotonicCalibrator":
        probs = softmax_numpy(logits)
        results = Parallel(n_jobs=-1, prefer="threads")(
            delayed(_fit_one_isotonic)(i, probs, y_true) for i in range(self.num_classes)
        )
        results.sort(key=lambda r: r[0])
        self.models = [r[1] for r in results]
        self.constant_values = [r[2] for r in results]
        return self

    def predict_proba(self, logits: np.ndarray) -> np.ndarray:
        probs = softmax_numpy(logits)
        calibrated = np.zeros_like(probs, dtype=np.float32)
        for i in range(self.num_classes):
            if self.models[i] is None:
                calibrated[:, i] = self.constant_values[i]
            else:
                calibrated[:, i] = self.models[i].transform(probs[:, i]).astype(np.float32)
        calibrated = np.clip(calibrated, EPS, 1.0)
        row_sums = np.sum(calibrated, axis=1, keepdims=True)
        row_sums = np.where(row_sums <= 0.0, 1.0, row_sums)
        return np.clip(calibrated / row_sums, EPS, 1.0)

    def save(self, path_prefix: Path) -> None:
        path_prefix.parent.mkdir(parents=True, exist_ok=True)
        payload = {"method": self.method_name, "num_classes": int(self.num_classes), "constant_values": self.constant_values}
        with (path_prefix.with_suffix(".json")).open("w", encoding="utf-8") as handle:
            json.dump(payload, handle, indent=2)
        joblib.dump(self.models, str(path_prefix.with_suffix(".joblib")))


# Kalibrátorok illesztése

def fit_all_calibrators(
    val_logits: np.ndarray, y_val: np.ndarray, num_classes: int, config: Dict[str, Any],
) -> OrderedDict:
    calibrators: OrderedDict[str, BaseCalibrator] = OrderedDict()
    calibrators["uncalibrated"] = IdentityCalibrator().fit(val_logits, y_val)

    specs = [
        ("temperature_scaling", lambda: TemperatureScalingCalibrator(
            steps=config["calibration_steps"], learning_rate=config["calibration_lr"])),
        ("static_temperature", lambda: StaticTemperatureCalibrator(temperature=1.5)),
        ("vector_scaling", lambda: VectorScalingCalibrator(
            num_classes=num_classes, steps=config["calibration_steps"],
            learning_rate=config["calibration_lr"], l2=config["calibration_l2"])),
        ("matrix_scaling", lambda: MatrixScalingCalibrator(
            num_classes=num_classes, steps=config["calibration_steps"],
            learning_rate=config["calibration_lr"], l2=config["calibration_l2"])),
        ("isotonic_regression", lambda: FastIsotonicCalibrator(num_classes=num_classes)),
    ]
    for method_name, factory in specs:
        try:
            calibrators[method_name] = factory().fit(val_logits, y_val)
        except Exception as exc:
            print(f"A(z) {method_name} kalibráció nem sikerült: {exc}. Kalibrálatlan kimenettel megy tovább.")
            calibrators[method_name] = IdentityCalibrator().fit(val_logits, y_val)
    return calibrators


def _cv_steps(config: Dict[str, Any], method: str) -> int:
    full = config["calibration_steps"]
    return {
        "temperature_scaling": full,
        "vector_scaling": max(30, full // 5),
        "matrix_scaling": max(20, full // 8),
        "isotonic_regression": 0,
    }.get(method, full)


def _fit_cv_calibrator(
    method_name: str, logits: np.ndarray, y: np.ndarray,
    num_classes: int, config: Dict[str, Any],
) -> BaseCalibrator:
    try:
        if method_name == "uncalibrated":
            return IdentityCalibrator().fit(logits, y)
        if method_name == "temperature_scaling":
            return TemperatureScalingCalibrator(steps=_cv_steps(config, method_name), learning_rate=config["calibration_lr"]).fit(logits, y)
        if method_name == "vector_scaling":
            return VectorScalingCalibrator(num_classes=num_classes, steps=_cv_steps(config, method_name), learning_rate=config["calibration_lr"], l2=config["calibration_l2"]).fit(logits, y)
        if method_name == "matrix_scaling":
            return MatrixScalingCalibrator(num_classes=num_classes, steps=_cv_steps(config, method_name), learning_rate=config["calibration_lr"], l2=config["calibration_l2"]).fit(logits, y)
        if method_name == "static_temperature":
            return StaticTemperatureCalibrator(temperature=1.5).fit(logits, y)
        if method_name == "isotonic_regression":
            return FastIsotonicCalibrator(num_classes=num_classes).fit(logits, y)
    except Exception as e:
        print(f"A CV illesztés nem sikerült ennél: {method_name}. Hiba: {e}")
    return IdentityCalibrator().fit(logits, y)


def fit_all_calibrators_cv(
    val_logits: np.ndarray, y_val: np.ndarray, num_classes: int,
    config: Dict[str, Any], seed: int,
) -> Tuple[OrderedDict, Dict[str, Dict[str, float]]]:
    """
    Teljes val halmazon illeszt downstream használatra.
    CV split alapján gap becslést is készít csökkentett lépésszámmal.
    """
    full_calibrators = fit_all_calibrators(val_logits, y_val, num_classes, config)

    n = len(y_val)
    cal_train_idx, cal_test_idx = train_test_split(
        np.arange(n), test_size=0.5, stratify=y_val, random_state=seed,
    )
    tr_logits, tr_y = val_logits[cal_train_idx], y_val[cal_train_idx]
    te_logits, te_y = val_logits[cal_test_idx], y_val[cal_test_idx]

    cv_gaps: Dict[str, Dict[str, float]] = {}
    for method_name in full_calibrators:
        cv_cal = _fit_cv_calibrator(method_name, tr_logits, tr_y, num_classes, config)
        try:
            tr_probs = cv_cal.predict_proba(tr_logits)
            te_probs = cv_cal.predict_proba(te_logits)
            tr_ece, _ = compute_ece_and_mce(tr_probs, tr_y, config["calibration_bins"])
            te_ece, _ = compute_ece_and_mce(te_probs, te_y, config["calibration_bins"])
            tr_nll = compute_nll(tr_probs, tr_y)
            te_nll = compute_nll(te_probs, te_y)
        except Exception:
            tr_ece = te_ece = tr_nll = te_nll = float("nan")
        cv_gaps[method_name] = {
            "cal_train_ece": tr_ece, "cal_test_ece": te_ece, "cal_ece_gap": te_ece - tr_ece,
            "cal_train_nll": tr_nll, "cal_test_nll": te_nll, "cal_nll_gap": te_nll - tr_nll,
        }
    return full_calibrators, cv_gaps


# 6. Kalibrációs metrikák

def compute_entropy(probs: np.ndarray) -> np.ndarray:
    probs = np.clip(probs, EPS, 1.0)
    return -np.sum(probs * np.log(probs), axis=1)


def compute_nll(probs: np.ndarray, y_true: np.ndarray) -> float:
    probs = np.clip(probs, EPS, 1.0)
    return float(-np.mean(np.log(probs[np.arange(len(y_true)), y_true])))


def compute_brier(probs: np.ndarray, y_true: np.ndarray) -> float:
    one_hot = np.zeros_like(probs, dtype=np.float32)
    one_hot[np.arange(len(y_true)), y_true] = 1.0
    return float(np.mean(np.sum((probs - one_hot) ** 2, axis=1)))


def compute_ece_and_mce(probs: np.ndarray, y_true: np.ndarray, n_bins: int = 15) -> Tuple[float, float]:
    confidences = np.max(probs, axis=1)
    predictions = np.argmax(probs, axis=1)
    bins = np.linspace(0.0, 1.0, n_bins + 1)
    ece, mce, total = 0.0, 0.0, len(y_true)
    for i in range(n_bins):
        mask = (confidences >= bins[i]) & (confidences <= bins[i + 1]) if i == 0 else (confidences > bins[i]) & (confidences <= bins[i + 1])
        if np.any(mask):
            acc = np.mean(predictions[mask] == y_true[mask])
            conf = np.mean(confidences[mask])
            gap = abs(acc - conf)
            ece += gap * np.sum(mask) / total
            mce = max(mce, gap)
    return float(ece), float(mce)


def compute_sce(probs: np.ndarray, y_true: np.ndarray, n_bins: int = 15) -> float:
    num_classes = probs.shape[1]
    bins = np.linspace(0.0, 1.0, n_bins + 1)
    total = len(y_true)
    sce = 0.0
    for class_idx in range(num_classes):
        class_probs = probs[:, class_idx]
        class_true = (y_true == class_idx).astype(np.float32)
        for i in range(n_bins):
            mask = (class_probs >= bins[i]) & (class_probs <= bins[i + 1]) if i == 0 else (class_probs > bins[i]) & (class_probs <= bins[i + 1])
            if np.any(mask):
                sce += abs(np.mean(class_true[mask]) - np.mean(class_probs[mask])) * np.sum(mask) / (total * num_classes)
    return float(sce)


def compute_ace(probs: np.ndarray, y_true: np.ndarray, n_bins: int = 15) -> float:
    num_classes = probs.shape[1]
    total = len(y_true)
    ace = 0.0
    for class_idx in range(num_classes):
        class_probs = probs[:, class_idx]
        class_true = (y_true == class_idx).astype(np.float32)
        order = np.argsort(class_probs)
        sorted_probs = class_probs[order]
        sorted_true = class_true[order]
        for idx in np.array_split(np.arange(total), n_bins):
            if len(idx) == 0:
                continue
            ace += abs(np.mean(sorted_true[idx]) - np.mean(sorted_probs[idx])) * len(idx) / (total * num_classes)
    return float(ace)


def compute_tace(probs: np.ndarray, y_true: np.ndarray, n_bins: int = 15, threshold: float = 0.01) -> float:
    num_classes = probs.shape[1]
    total = len(y_true)
    tace = 0.0
    for class_idx in range(num_classes):
        class_probs = probs[:, class_idx]
        class_true = (y_true == class_idx).astype(np.float32)
        keep_mask = class_probs > threshold
        if not np.any(keep_mask):
            continue
        filtered_probs = class_probs[keep_mask]
        filtered_true = class_true[keep_mask]
        order = np.argsort(filtered_probs)
        sorted_probs = filtered_probs[order]
        sorted_true = filtered_true[order]
        for idx in np.array_split(np.arange(len(sorted_probs)), n_bins):
            if len(idx) == 0:
                continue
            tace += abs(np.mean(sorted_true[idx]) - np.mean(sorted_probs[idx])) * len(idx) / (total * num_classes)
    return float(tace)



def entropy_relationships(probs: np.ndarray, y_true: np.ndarray, n_bins: int = 15) -> Dict[str, np.ndarray]:
    entropy = compute_entropy(probs)
    edges = np.linspace(float(entropy.min()), float(entropy.max()) + 1e-8, n_bins + 1)
    centers = 0.5 * (edges[:-1] + edges[1:])
    accuracy = np.full(n_bins, np.nan, dtype=np.float32)
    error_rate = np.full(n_bins, np.nan, dtype=np.float32)
    predictions = np.argmax(probs, axis=1)
    for i in range(n_bins):
        mask = (entropy >= edges[i]) & (entropy <= edges[i + 1]) if i == n_bins - 1 else (entropy >= edges[i]) & (entropy < edges[i + 1])
        if np.any(mask):
            correct = predictions[mask] == y_true[mask]
            accuracy[i] = float(np.mean(correct))
            error_rate[i] = 1.0 - accuracy[i]
    return {"centers": centers, "accuracy": accuracy, "error_rate": error_rate, "entropy": entropy}


def selective_prediction_stats(
    probs: np.ndarray, y_true: np.ndarray,
    coverage_levels: List[float], risk_levels: List[float],
    use_entropy: bool = False,
) -> Dict[str, Any]:
    predictions = np.argmax(probs, axis=1)
    uncertainty = compute_entropy(probs) if use_entropy else 1.0 - np.max(probs, axis=1)
    order = np.argsort(uncertainty)
    sorted_errors = (predictions[order] != y_true[order]).astype(np.float32)
    accepted = np.arange(1, len(sorted_errors) + 1)
    coverage = accepted / len(sorted_errors)
    risk = np.cumsum(sorted_errors) / accepted
    aurc = float(np.trapezoid(risk, coverage))

    risk_at_cov = {}
    for cov in coverage_levels:
        idx = int(np.clip(np.ceil(cov * len(coverage)) - 1, 0, len(coverage) - 1))
        risk_at_cov[f"risk_at_cov_{int(cov * 100)}"] = float(risk[idx])

    cov_at_risk = {}
    for risk_threshold in risk_levels:
        valid = np.where(risk <= risk_threshold)[0]
        cov_at_risk[f"coverage_at_risk_{int(risk_threshold * 100)}"] = float(coverage[valid[-1]]) if len(valid) > 0 else 0.0

    return {"coverage": coverage, "risk": risk, "AURC": aurc, **risk_at_cov, **cov_at_risk}


def evaluate_probabilities(
    probs: np.ndarray, y_true: np.ndarray, n_bins: int,
    tace_threshold: float, coverage_levels: List[float], risk_levels: List[float],
) -> Tuple[Dict[str, float], Dict[str, Any]]:
    probs = np.clip(np.asarray(probs, dtype=np.float32), EPS, 1.0)
    probs = probs / np.sum(probs, axis=1, keepdims=True)
    predictions = np.argmax(probs, axis=1)

    accuracy = float(np.mean(predictions == y_true))
    macro_precision, macro_recall, macro_f1, _ = precision_recall_fscore_support(y_true, predictions, average="macro", zero_division=0)
    weighted_f1 = precision_recall_fscore_support(y_true, predictions, average="weighted", zero_division=0)[2]
    per_class_precision, per_class_recall, per_class_f1, _ = precision_recall_fscore_support(y_true, predictions, average=None, zero_division=0)

    mcc = float(matthews_corrcoef(y_true, predictions))
    kappa = float(cohen_kappa_score(y_true, predictions))
    balanced_acc = float(balanced_accuracy_score(y_true, predictions))

    nll = compute_nll(probs, y_true)
    brier = compute_brier(probs, y_true)
    ece, mce = compute_ece_and_mce(probs, y_true, n_bins=n_bins)
    ace = compute_ace(probs, y_true, n_bins=n_bins)
    sce = compute_sce(probs, y_true, n_bins=n_bins)
    tace = compute_tace(probs, y_true, n_bins=n_bins, threshold=tace_threshold)

    confidence = np.max(probs, axis=1)
    entropy = compute_entropy(probs)
    confidence_percentiles = np.percentile(confidence, [5, 25, 50, 75, 95]).astype(np.float32)
    entropy_percentiles = np.percentile(entropy, [5, 25, 50, 75, 95]).astype(np.float32)

    selective = selective_prediction_stats(probs, y_true, coverage_levels, risk_levels, use_entropy=False)
    entropy_curves = entropy_relationships(probs, y_true, n_bins=n_bins)
    cm = confusion_matrix(y_true, predictions)

    metrics = {
        "accuracy": accuracy, "balanced_accuracy": balanced_acc,
        "macro_precision": float(macro_precision), "macro_recall": float(macro_recall),
        "macro_F1": float(macro_f1), "weighted_F1": float(weighted_f1),
        "MCC": mcc, "cohen_kappa": kappa,
        "per_class_F1_mean": float(np.mean(per_class_f1)), "per_class_F1_std": float(np.std(per_class_f1)),
        "per_class_F1_min": float(np.min(per_class_f1)), "per_class_F1_max": float(np.max(per_class_f1)),
        "NLL": nll, "Brier": brier, "ECE": ece, "ACE": ace, "SCE": sce, "TACE": tace, "MCE": mce,
        "entropy_mean": float(np.mean(entropy)), "entropy_std": float(np.std(entropy)),
        "confidence_mean": float(np.mean(confidence)),
        "confidence_accuracy_gap": float(np.mean(confidence) - accuracy),
        "confidence_p5": float(confidence_percentiles[0]), "confidence_p25": float(confidence_percentiles[1]),
        "confidence_p50": float(confidence_percentiles[2]), "confidence_p75": float(confidence_percentiles[3]),
        "confidence_p95": float(confidence_percentiles[4]),
        "entropy_p5": float(entropy_percentiles[0]), "entropy_p25": float(entropy_percentiles[1]),
        "entropy_p50": float(entropy_percentiles[2]), "entropy_p75": float(entropy_percentiles[3]),
        "entropy_p95": float(entropy_percentiles[4]),
        "AURC": selective["AURC"],
        "risk_at_cov_80": selective.get("risk_at_cov_80", np.nan),
        "risk_at_cov_90": selective.get("risk_at_cov_90", np.nan),
        "risk_at_cov_95": selective.get("risk_at_cov_95", np.nan),
        "coverage_at_risk_5": selective.get("coverage_at_risk_5", np.nan),
        "coverage_at_risk_10": selective.get("coverage_at_risk_10", np.nan),
        "coverage_at_risk_20": selective.get("coverage_at_risk_20", np.nan),
    }
    aux = {
        "predictions": predictions, "confidence": confidence, "entropy": entropy,
        "confusion_matrix": cm, "coverage": selective["coverage"], "risk": selective["risk"],
        "entropy_curve_centers": entropy_curves["centers"],
        "entropy_curve_accuracy": entropy_curves["accuracy"],
        "entropy_curve_error_rate": entropy_curves["error_rate"],
        "per_class_f1": per_class_f1, "per_class_precision": per_class_precision,
        "per_class_recall": per_class_recall,
    }
    return metrics, aux


# Pszeudocímkék generálása

def generate_pseudo_labels(
    x_unlabeled: np.ndarray, y_unlabeled_true: Optional[np.ndarray],
    probs: np.ndarray, num_classes: int,
    confidence_threshold: float = 0.9, entropy_percentile: float = 30.0,
    iteration: int = 0,
    confidence_threshold_increment: float = 0.02, entropy_percentile_decrement: float = 5.0,
) -> Tuple[np.ndarray, np.ndarray, np.ndarray, Dict[str, Any], np.ndarray, np.ndarray]:
    probs = np.asarray(probs, dtype=np.float32)
    conf = np.max(probs, axis=1)
    entropy = compute_entropy(probs)
    pred = np.argmax(probs, axis=1).astype(np.int32)

    effective_confidence = min(confidence_threshold + iteration * confidence_threshold_increment, 0.99)
    effective_entropy_pct = max(entropy_percentile - iteration * entropy_percentile_decrement, 5.0)
    entropy_threshold = float(np.percentile(entropy, effective_entropy_pct))
    accepted_mask = (conf >= effective_confidence) & (entropy <= entropy_threshold)
    accepted_indices = np.where(accepted_mask)[0]

    x_pseudo = x_unlabeled[accepted_indices]
    y_pseudo = pred[accepted_indices]

    accepted_count = int(len(accepted_indices))
    total_count = int(len(x_unlabeled))
    acceptance_ratio = float(accepted_count / total_count) if total_count > 0 else 0.0

    pseudo_noise_rate = np.nan
    if y_unlabeled_true is not None and accepted_count > 0:
        pseudo_noise_rate = float(np.mean(y_pseudo != y_unlabeled_true[accepted_indices]))

    class_distribution = np.bincount(y_pseudo, minlength=num_classes).astype(int).tolist() if accepted_count > 0 else [0] * num_classes

    stats = {
        "entropy_threshold": entropy_threshold,
        "effective_confidence_threshold": effective_confidence,
        "effective_entropy_percentile": effective_entropy_pct,
        "acceptance_ratio": acceptance_ratio,
        "accepted_count": accepted_count,
        "rejected_count": total_count - accepted_count,
        "pseudo_noise_rate": pseudo_noise_rate,
        "class_distribution_json": json.dumps(class_distribution),
        "accepted_confidence_mean": float(np.mean(conf[accepted_mask])) if accepted_count > 0 else np.nan,
        "accepted_entropy_mean": float(np.mean(entropy[accepted_mask])) if accepted_count > 0 else np.nan,
        "accepted_confidence_p5": float(np.percentile(conf[accepted_mask], 5)) if accepted_count > 0 else np.nan,
        "accepted_confidence_p95": float(np.percentile(conf[accepted_mask], 95)) if accepted_count > 0 else np.nan,
        "accepted_entropy_p5": float(np.percentile(entropy[accepted_mask], 5)) if accepted_count > 0 else np.nan,
        "accepted_entropy_p95": float(np.percentile(entropy[accepted_mask], 95)) if accepted_count > 0 else np.nan,
    }
    return x_pseudo, y_pseudo, accepted_mask, stats, conf, entropy


def concat_features(x_a: np.ndarray, x_b: np.ndarray) -> np.ndarray:
    if len(x_b) == 0:
        return x_a.copy()
    return np.concatenate([x_a, x_b], axis=0)



def build_dataset_output_dirs(base_output_dir: Path, dataset_name: str) -> Dict[str, Path]:
    dataset_root = base_output_dir / dataset_name
    dirs = {
        "root": dataset_root,
        "checkpoints": dataset_root / "checkpoints",
        "histories": dataset_root / "histories",
        "metrics": dataset_root / "metrics",
        "pseudo_labels": dataset_root / "pseudo_labels",
        "summaries": dataset_root / "summaries",
    }
    for d in dirs.values():
        d.mkdir(parents=True, exist_ok=True)
    return dirs


def _history_proxy(history_payload: Dict[str, Any]) -> Any:
    return type("HistoryProxy", (), {"history": history_payload})()


def _ratio_token(labeled_ratio: float) -> str:
    return f"ratio{int(round(labeled_ratio * 100))}"


def run_single_configuration(
    bundle: DatasetBundle, labeled_ratio: float, seed: int,
    config: Dict[str, Any], dataset_dirs: Dict[str, Path],
    capture_reference: bool = False,
) -> Tuple[List, List, List, Dict, List]:

    set_global_seed(seed)

    x_train, y_train = maybe_subsample_stratified(
        bundle.train_x, bundle.train_y, config["max_train_samples"], seed,
    )
    x_test, y_test = maybe_subsample_stratified(
        bundle.test_x, bundle.test_y, config["max_test_samples"], seed,
    )

    train_idx, val_idx = train_test_split(
        np.arange(len(y_train)), test_size=config["val_ratio"],
        stratify=y_train, random_state=seed,
    )
    x_train_pool = x_train[train_idx]
    y_train_pool = y_train[train_idx]
    x_val = x_train[val_idx]
    y_val = y_train[val_idx]

    labeled_idx, unlabeled_idx = train_test_split(
        np.arange(len(y_train_pool)), train_size=labeled_ratio,
        stratify=y_train_pool, random_state=seed,
    )
    x_labeled = x_train_pool[labeled_idx]
    y_labeled = y_train_pool[labeled_idx]
    x_unlabeled_full = x_train_pool[unlabeled_idx]
    y_unlabeled_full = y_train_pool[unlabeled_idx]

    print(
        f"{bundle.name}: ratio={labeled_ratio:.2f}, seed={seed}, "
        f"labeled={len(y_labeled)}, unlabeled={len(x_unlabeled_full)}, "
        f"val={len(y_val)}, test={len(y_test)}"
    )

    # A validációs és teszt halmaznál a resize csak egyszer fusson le.
    val_ds = make_labeled_dataset(
        x=x_val, y=y_val, batch_size=config["batch_size"],
        is_path_dataset=bundle.is_path_dataset, image_size=config["image_size"],
        training=False, seed=seed, use_cache=True,
    )
    test_ds = make_labeled_dataset(
        x=x_test, y=y_test, batch_size=config["batch_size"],
        is_path_dataset=bundle.is_path_dataset, image_size=config["image_size"],
        training=False, seed=seed, use_cache=True,
    )

    results_rows: List[Dict] = []
    pseudo_rows: List[Dict] = []
    gap_rows: List[Dict] = []
    cal_cv_rows: List[Dict] = []
    reference_artifacts: Dict = {}

    # Kumulatív tanítóhalmaz és a folyamatosan szűkülő címkézetlen pool.
    current_train_x = x_labeled.copy()
    current_train_y = y_labeled.copy()
    remaining_unl_idx = np.arange(len(x_unlabeled_full))

    ratio_token = _ratio_token(labeled_ratio)

    # Ha van korábbi állapot, azt visszatöltjük.
    first_unfinished_iteration = 0
    for prev_iter in range(config["teacher_student_rounds"]):
        prev_name = f"{bundle.name}_{ratio_token}_seed{seed}_iter{prev_iter}"
        prev_done = dataset_dirs["checkpoints"] / f"{prev_name}_done.flag"
        cumul_x_path = dataset_dirs["pseudo_labels"] / f"{prev_name}_cumulative_train_x.npy"
        cumul_y_path = dataset_dirs["pseudo_labels"] / f"{prev_name}_cumulative_train_y.npy"
        remain_path = dataset_dirs["pseudo_labels"] / f"{prev_name}_remaining_unl_idx.npy"
        if not prev_done.exists():
            break
        if not (cumul_x_path.exists() and cumul_y_path.exists() and remain_path.exists()):
            print(
                f"  Az {prev_iter}. iteráció késznek van jelölve, de hiányzik néhány mentett állapot. "
                f"Innen újrafuttatom."
            )
            prev_done.unlink()
            break
        current_train_x = np.load(str(cumul_x_path), allow_pickle=True)
        current_train_y = np.load(str(cumul_y_path), allow_pickle=True)
        remaining_unl_idx = np.load(str(remain_path), allow_pickle=True)
        first_unfinished_iteration = prev_iter + 1
        print(
            f"  {prev_iter}. iteráció állapota visszatöltve: "
            f"train={len(current_train_y)}, remaining_unlabeled={len(remaining_unl_idx)}"
        )

    for iteration in range(config["teacher_student_rounds"]):
        run_name = f"{bundle.name}_{ratio_token}_seed{seed}_iter{iteration}"
        if iteration < first_unfinished_iteration:
            round_csv = dataset_dirs["metrics"] / f"results_seed{seed}_{ratio_token}_round{iteration + 1}.csv"
            if round_csv.exists():
                results_rows.extend(pd.read_csv(round_csv).to_dict("records"))
            pseudo_csv = dataset_dirs["pseudo_labels"] / f"pseudo_stats_seed{seed}_{ratio_token}_round{iteration + 1}.csv"
            if pseudo_csv.exists():
                pseudo_rows.extend(pd.read_csv(pseudo_csv).to_dict("records"))
            history_path = dataset_dirs["histories"] / f"{run_name}_history.json"
            if history_path.exists():
                with history_path.open("r", encoding="utf-8") as _f:
                    hp = json.load(_f)
                role = "teacher" if iteration == 0 else "student"
                gap_stats = extract_generalization_gap(_history_proxy(hp))
                gap_rows.append({
                    "dataset": bundle.name, "labeled_ratio": labeled_ratio,
                    "iteration": iteration, "seed": seed, "role": role, **gap_stats
                })
                if capture_reference:
                    if iteration == 0:
                        reference_artifacts["teacher_history"] = hp
                    if iteration == config["teacher_student_rounds"] - 1:
                        reference_artifacts["student_history"] = hp
            print(f"  A(z) {iteration}. iteráció már kész volt, a mentett eredményeket töltöm be.")
            continue

        tf.keras.backend.clear_session()
        gc.collect()
        set_global_seed(seed + iteration)

        checkpoint_path = dataset_dirs["checkpoints"] / f"{run_name}_best.weights.h5"
        final_wt_path = dataset_dirs["root"] / f"{run_name}_final.weights.h5"
        history_path = dataset_dirs["histories"] / f"{run_name}_history.json"
        done_marker = dataset_dirs["checkpoints"] / f"{run_name}_done.flag"

        augmentation = build_data_augmentation(seed + iteration)
        train_ds = make_labeled_dataset(
            x=current_train_x, y=current_train_y,
            batch_size=config["batch_size"],
            is_path_dataset=bundle.is_path_dataset, image_size=config["image_size"],
            training=True, seed=seed + iteration, augmentation=augmentation,
            use_cache=False,
        )

        model = build_classifier(
            num_classes=bundle.num_classes, image_size=config["image_size"],
            learning_rate=config["learning_rate"],
            backbone_trainable=(iteration > 0),
        )

        # Tanítás vagy egy korábban elmentett állapot visszatöltése.
        if checkpoint_path.exists() and history_path.exists():
            print(f"  {run_name}: a tanítás már kész, a súlyokat betöltöm.")
            model.load_weights(str(checkpoint_path))
            with history_path.open("r", encoding="utf-8") as _f:
                history_payload = json.load(_f)
        else:
            if checkpoint_path.exists():
                print(f"  Félbeszakadt tanítás, újrakezdem: {run_name}")
                checkpoint_path.unlink()
            history = train_model(
                model=model, train_ds=train_ds, val_ds=val_ds,
                epochs=config["epochs"], checkpoint_path=checkpoint_path,
                early_stop_patience=config["early_stop_patience"],
                reduce_lr_patience=config["reduce_lr_patience"],
            )
            save_history(history, history_path)
            history_payload = history.history

        model.save_weights(str(final_wt_path))

        role = "teacher" if iteration == 0 else "student"
        gap_stats = extract_generalization_gap(_history_proxy(history_payload))
        gap_rows.append({
            "dataset": bundle.name, "labeled_ratio": labeled_ratio,
            "iteration": iteration, "seed": seed, "role": role, **gap_stats
        })

        val_logits = predict_logits(model, val_ds)
        calibrators, cv_gaps = fit_all_calibrators_cv(
            val_logits, y_val, bundle.num_classes, config, seed,
        )
        for method_name, gap_dict in cv_gaps.items():
            cal_cv_rows.append({
                "dataset": bundle.name, "labeled_ratio": labeled_ratio,
                "iteration": iteration, "seed": seed,
                "calibration_method": method_name, **gap_dict
            })
        for method_name, calibrator in calibrators.items():
            calibrator.save(dataset_dirs["checkpoints"] / f"{run_name}_{method_name}")

        test_logits = predict_logits(model, test_ds)

        # A megmaradt címkézetlen mintákon generálunk pszeudocímkéket.
        x_remaining = x_unlabeled_full[remaining_unl_idx]
        y_remaining = y_unlabeled_full[remaining_unl_idx]
        unlabeled_ds = make_unlabeled_dataset(
            x=x_remaining, batch_size=config["batch_size"],
            is_path_dataset=bundle.is_path_dataset, image_size=config["image_size"],
        )
        unlabeled_logits = predict_logits(model, unlabeled_ds)

        pseudo_method = config["pseudo_label_calibration_method"]
        pseudo_probs = calibrators[pseudo_method].predict_proba(unlabeled_logits)

        x_pseudo, y_pseudo, accepted_mask, p_stats, unl_conf, unl_entropy = generate_pseudo_labels(
            x_unlabeled=x_remaining, y_unlabeled_true=y_remaining,
            probs=pseudo_probs, num_classes=bundle.num_classes,
            confidence_threshold=config["confidence_threshold"],
            entropy_percentile=config["entropy_percentile"],
            iteration=iteration,
            confidence_threshold_increment=config["confidence_threshold_increment"],
            entropy_percentile_decrement=config["entropy_percentile_decrement"],
        )

        # A kiválasztott pszeudocímkéket hozzáadjuk a tanítóhalmazhoz.
        if len(y_pseudo) > 0:
            current_train_x = np.concatenate([current_train_x, x_pseudo], axis=0)
            current_train_y = np.concatenate([current_train_y, y_pseudo], axis=0)
            remaining_unl_idx = np.delete(remaining_unl_idx, np.where(accepted_mask)[0])
            print(
                f"  iter {iteration}: elfogadott={len(y_pseudo)}, "
                f"zaj={p_stats['pseudo_noise_rate']:.4f}, "
                f"train={len(current_train_y)}, "
                f"remaining_unlabeled={len(remaining_unl_idx)}"
            )
        else:
            print(f"  iter {iteration}: nem lett elfogadott pszeudocímke.")

        # Mentés
        pred_remaining = np.argmax(pseudo_probs, axis=1)
        eff_conf_thr = p_stats["effective_confidence_threshold"]

        no_filter_noise = float(np.mean(pred_remaining != y_remaining)) if len(y_remaining) > 0 else np.nan
        no_filter_count = len(y_remaining)

        conf_only_mask = unl_conf >= eff_conf_thr
        conf_only_count = int(np.sum(conf_only_mask))
        conf_only_noise = float(np.mean(pred_remaining[conf_only_mask] != y_remaining[conf_only_mask])) if conf_only_count > 0 else np.nan
        conf_only_ratio = float(conf_only_count / len(y_remaining)) if len(y_remaining) > 0 else 0.0

        pseudo_row = {
            "dataset": bundle.name, "labeled_ratio": labeled_ratio,
            "iteration": iteration, "seed": seed,
            "calibration_for_pseudo_labels": pseudo_method,
            "confidence_threshold": config["confidence_threshold"],
            "entropy_percentile": config["entropy_percentile"],
            **p_stats,
            "no_filter_noise_rate": no_filter_noise,
            "no_filter_accepted_count": no_filter_count,
            "confidence_only_noise_rate": conf_only_noise,
            "confidence_only_accepted_count": conf_only_count,
            "confidence_only_acceptance_ratio": conf_only_ratio,
        }
        pseudo_rows.append(pseudo_row)
        pd.DataFrame([pseudo_row]).to_csv(
            dataset_dirs["pseudo_labels"] / f"pseudo_stats_seed{seed}_{ratio_token}_round{iteration + 1}.csv",
            index=False,
        )
        np.save(dataset_dirs["pseudo_labels"] / f"{run_name}_pseudo_accept_mask.npy", accepted_mask.astype(np.int8))
        np.save(dataset_dirs["pseudo_labels"] / f"{run_name}_pseudo_confidence.npy", unl_conf.astype(np.float32))
        np.save(dataset_dirs["pseudo_labels"] / f"{run_name}_pseudo_entropy.npy", unl_entropy.astype(np.float32))
        np.save(dataset_dirs["pseudo_labels"] / f"{run_name}_pseudo_labels.npy", y_pseudo.astype(np.int32))
        np.save(dataset_dirs["pseudo_labels"] / f"{run_name}_remaining_unl_idx.npy", remaining_unl_idx)
        np.save(dataset_dirs["pseudo_labels"] / f"{run_name}_cumulative_train_x.npy", current_train_x)
        np.save(dataset_dirs["pseudo_labels"] / f"{run_name}_cumulative_train_y.npy", current_train_y)

        # Teszt metrikák
        round_metrics_rows: List[Dict] = []
        for method_name, calibrator in calibrators.items():
            probs = calibrator.predict_proba(test_logits)
            metrics, aux = evaluate_probabilities(
                probs=probs, y_true=y_test,
                n_bins=config["calibration_bins"], tace_threshold=config["tace_threshold"],
                coverage_levels=config["coverage_levels"], risk_levels=config["risk_levels"],
            )
            row = {
                "dataset": bundle.name, "labeled_ratio": labeled_ratio,
                "iteration": iteration, "calibration_method": method_name, "seed": seed,
                "pseudo_label_acceptance_ratio": p_stats["acceptance_ratio"],
                "training_set_size": len(current_train_y),
                **metrics,
            }
            round_metrics_rows.append(row)
            results_rows.append(row)

            if capture_reference and iteration == config["teacher_student_rounds"] - 1:
                reference_artifacts.setdefault("final_probs_by_method", {})[method_name] = probs
                reference_artifacts["final_y_test"] = y_test
                reference_artifacts.setdefault("final_confusion_by_method", {})[method_name] = aux["confusion_matrix"]
                reference_artifacts.setdefault("final_coverage_by_method", {})[method_name] = aux["coverage"]
                reference_artifacts.setdefault("final_risk_by_method", {})[method_name] = aux["risk"]
                reference_artifacts["final_entropy_curve"] = {
                    "centers": aux["entropy_curve_centers"],
                    "accuracy": aux["entropy_curve_accuracy"],
                    "error_rate": aux["entropy_curve_error_rate"],
                }
                reference_artifacts["final_confidence_distribution"] = aux["confidence"]
                reference_artifacts["final_entropy_distribution"] = aux["entropy"]

        pd.DataFrame(round_metrics_rows).to_csv(
            dataset_dirs["metrics"] / f"results_seed{seed}_{ratio_token}_round{iteration + 1}.csv",
            index=False,
        )

        if capture_reference:
            if iteration == 0:
                reference_artifacts["teacher_history"] = history_payload
            if iteration == config["teacher_student_rounds"] - 1:
                reference_artifacts["student_history"] = history_payload
            reference_artifacts["pseudo_confidence_all"] = unl_conf
            reference_artifacts["pseudo_entropy_all"] = unl_entropy
            reference_artifacts["pseudo_accept_mask"] = accepted_mask

        # A done jelzőt csak minden mentés után írjuk ki.
        done_marker.write_text("done")

        del train_ds, model, val_logits, test_logits, unlabeled_logits, calibrators
        gc.collect()

    config_done_marker = dataset_dirs["summaries"] / f"done_{ratio_token}_seed{seed}.flag"
    config_done_marker.parent.mkdir(parents=True, exist_ok=True)
    config_done_marker.write_text("done")

    return results_rows, pseudo_rows, gap_rows, reference_artifacts, cal_cv_rows


def _load_config_results_from_disk(
    dataset_dirs: Dict[str, Path], dataset_name: str,
    labeled_ratio: float, seed: int, config: Dict[str, Any],
) -> Tuple[List[Dict], List[Dict], List[Dict], List[Dict]]:
    """Betölti a korábban lementett körönkénti CSV-ket egy befejezett konfigurációhoz."""
    ratio_token = _ratio_token(labeled_ratio)
    results_rows: List[Dict] = []
    pseudo_rows: List[Dict] = []
    gap_rows: List[Dict] = []
    cal_cv_rows: List[Dict] = []
    for iteration in range(config["teacher_student_rounds"]):
        metrics_csv = dataset_dirs["metrics"] / f"results_seed{seed}_{ratio_token}_round{iteration + 1}.csv"
        if metrics_csv.exists():
            results_rows.extend(pd.read_csv(metrics_csv).to_dict("records"))
        pseudo_csv = dataset_dirs["pseudo_labels"] / f"pseudo_stats_seed{seed}_{ratio_token}_round{iteration + 1}.csv"
        if pseudo_csv.exists():
            pseudo_rows.extend(pd.read_csv(pseudo_csv).to_dict("records"))
    return results_rows, pseudo_rows, gap_rows, cal_cv_rows


def run_all_experiments(config: Dict[str, Any]) -> Tuple[pd.DataFrame, pd.DataFrame, pd.DataFrame, Dict[str, Any], pd.DataFrame]:
    all_results: List[Dict] = []
    all_pseudo: List[Dict] = []
    all_gaps: List[Dict] = []
    all_cal_cv: List[Dict] = []
    reference_artifacts: Dict = {}

    datasets = config["datasets"]
    seeds = config["seeds"]
    labeled_ratios = config["labeled_ratios"]

    reference_dataset = datasets[0]
    reference_ratio = labeled_ratios[len(labeled_ratios) // 2]
    reference_seed = seeds[0]

    for dataset_name in datasets:
        dataset_dirs = build_dataset_output_dirs(config["output_dir"], dataset_name)
        dataset_results: List[Dict] = []
        dataset_pseudo: List[Dict] = []
        dataset_gaps: List[Dict] = []

        all_configs_done = all(
            (dataset_dirs["summaries"] / f"done_{_ratio_token(lr)}_seed{s}.flag").exists()
            for lr in labeled_ratios for s in seeds
        )

        bundle = None
        if not all_configs_done:
            try:
                bundle = load_dataset_bundle(dataset_name)
            except Exception as exc:
                message = f"Nem sikerült betölteni ezt az adathalmazt: '{dataset_name}'. Hiba: {exc}"
                if config["strict_dataset_loading"]:
                    raise RuntimeError(message) from exc
                print(message)
                continue
        else:
            print(f"{dataset_name}: minden konfiguráció kész, az adatbetöltést kihagyom.")

        for labeled_ratio in labeled_ratios:
            for seed in seeds:
                ratio_token = _ratio_token(labeled_ratio)
                config_done_marker = dataset_dirs["summaries"] / f"done_{ratio_token}_seed{seed}.flag"

                if config_done_marker.exists():
                    print(f"{dataset_name}: ratio={labeled_ratio}, seed={seed} már kész, betöltöm a mentett eredményeket.")
                    res, pse, gap, ccv = _load_config_results_from_disk(
                        dataset_dirs, dataset_name, labeled_ratio, seed, config,
                    )
                    all_results.extend(res)
                    all_pseudo.extend(pse)
                    all_gaps.extend(gap)
                    all_cal_cv.extend(ccv)
                    dataset_results.extend(res)
                    dataset_pseudo.extend(pse)
                    dataset_gaps.extend(gap)
                    continue

                capture_reference = (
                    dataset_name == reference_dataset
                    and float(labeled_ratio) == float(reference_ratio)
                    and seed == reference_seed
                )
                results, pseudo, gaps, ref, cal_cv = run_single_configuration(
                    bundle=bundle, labeled_ratio=float(labeled_ratio), seed=int(seed),
                    config=config, dataset_dirs=dataset_dirs, capture_reference=capture_reference,
                )
                all_results.extend(results)
                all_pseudo.extend(pseudo)
                all_gaps.extend(gaps)
                all_cal_cv.extend(cal_cv)
                dataset_results.extend(results)
                dataset_pseudo.extend(pseudo)
                dataset_gaps.extend(gaps)
                if capture_reference:
                    reference_artifacts = ref

                pd.DataFrame(dataset_results).to_csv(dataset_dirs["summaries"] / "summary_results.csv", index=False)
                pd.DataFrame(dataset_pseudo).to_csv(dataset_dirs["summaries"] / "summary_pseudo_stats.csv", index=False)
                pd.DataFrame(dataset_gaps).to_csv(dataset_dirs["summaries"] / "summary_generalization_gaps.csv", index=False)

    return pd.DataFrame(all_results), pd.DataFrame(all_pseudo), pd.DataFrame(all_gaps), reference_artifacts, pd.DataFrame(all_cal_cv)


# Ábrák mentése

def save_and_show_plot(path: Path) -> None:
    path.parent.mkdir(parents=True, exist_ok=True)
    plt.tight_layout()
    plt.savefig(path, dpi=180)
    try:
        plt.show()
    except Exception:
        pass
    plt.close()


# Statisztikai táblák és aggregálás

REQUIRED_RESULT_COLUMNS = [
    "dataset", "labeled_ratio", "iteration", "calibration_method", "seed",
    "accuracy", "balanced_accuracy", "macro_precision", "macro_recall",
    "macro_F1", "weighted_F1", "MCC", "cohen_kappa",
    "per_class_F1_mean", "per_class_F1_std", "per_class_F1_min", "per_class_F1_max",
    "NLL", "Brier", "ECE", "ACE", "SCE", "TACE", "MCE",
    "entropy_mean", "entropy_std", "confidence_mean", "confidence_accuracy_gap",
    "AURC", "pseudo_label_acceptance_ratio",
]


def aggregate_mean_std(df: pd.DataFrame, group_cols: List[str], metric_cols: List[str]) -> pd.DataFrame:
    mean_df = df.groupby(group_cols, as_index=False)[metric_cols].mean()
    std_df = df.groupby(group_cols, as_index=False)[metric_cols].std(ddof=1)
    std_df = std_df.rename(columns={col: f"{col}_std" for col in metric_cols})
    count_df = df.groupby(group_cols, as_index=False)[metric_cols[0]].count()
    count_df = count_df.rename(columns={metric_cols[0]: "_n_seeds"})

    merged = mean_df.merge(std_df, on=group_cols, how="left").merge(count_df, on=group_cols, how="left")

    for col in metric_cols:
        merged[f"{col}_mean"] = merged[col]
        n = merged["_n_seeds"]
        std_col = merged[f"{col}_std"]
        t_crit = np.array([
            scipy_stats.t.ppf(0.975, max(ni - 1, 1)) if ni > 1 else np.nan for ni in n
        ])
        ci_half = t_crit * std_col / np.sqrt(np.maximum(n, 1))
        merged[f"{col}_ci95"] = ci_half
        merged[f"{col}_ci95_low"] = merged[col] - ci_half
        merged[f"{col}_ci95_high"] = merged[col] + ci_half
        merged[f"{col}_mean_std"] = merged.apply(lambda row, c=col: f"{row[c]:.4f} +- {row[f'{c}_std']:.4f}", axis=1)
        merged[f"{col}_mean_ci"] = merged.apply(lambda row, c=col: f"{row[c]:.4f} [{row[f'{c}_ci95_low']:.4f}, {row[f'{c}_ci95_high']:.4f}]", axis=1)
    return merged


def compute_pairwise_significance(df: pd.DataFrame, method_col: str, metric_cols: List[str], alpha: float = 0.05) -> pd.DataFrame:
    methods = sorted(df[method_col].unique())
    method_pairs = list(combinations(methods, 2))
    n_comparisons = len(method_pairs)
    corrected_alpha = alpha / max(n_comparisons, 1)
    rows = []
    for m1, m2 in method_pairs:
        d1 = df[df[method_col] == m1].sort_values("seed")
        d2 = df[df[method_col] == m2].sort_values("seed")
        if len(d1) != len(d2) or len(d1) < 2:
            continue
        for metric in metric_cols:
            v1, v2 = d1[metric].values, d2[metric].values
            valid = ~(np.isnan(v1) | np.isnan(v2))
            if valid.sum() < 2:
                continue
            t_stat, p_value = scipy_stats.ttest_rel(v1[valid], v2[valid])
            diff = v1[valid] - v2[valid]
            d_cohen = float(np.mean(diff) / max(np.std(diff, ddof=1), EPS))
            rows.append({
                "method_1": m1, "method_2": m2, "metric": metric,
                "t_statistic": float(t_stat), "p_value": float(p_value),
                "p_value_bonferroni": float(min(p_value * n_comparisons, 1.0)),
                "significant_raw": p_value < alpha,
                "significant_bonferroni": p_value < corrected_alpha,
                "cohens_d": d_cohen, "mean_diff": float(np.mean(diff)),
                "n_comparisons": n_comparisons, "corrected_alpha": corrected_alpha,
            })
    return pd.DataFrame(rows)


def compute_friedman_test(df: pd.DataFrame, method_col: str, metric_cols: List[str]) -> pd.DataFrame:
    methods = sorted(df[method_col].unique())
    seeds = sorted(df["seed"].unique())
    rows = []
    for metric in metric_cols:
        groups = []
        for method in methods:
            values = df[df[method_col] == method].sort_values("seed")[metric].values
            if len(values) == len(seeds):
                groups.append(values)
        if len(groups) < 3:
            continue
        try:
            stat, p_value = scipy_stats.friedmanchisquare(*groups)
            rows.append({
                "metric": metric, "friedman_statistic": float(stat), "p_value": float(p_value),
                "n_methods": len(groups), "n_seeds": len(seeds),
                "significant_005": p_value < 0.05, "significant_001": p_value < 0.01
            })
        except Exception:
            pass
    return pd.DataFrame(rows)


def compute_iteration_significance(df: pd.DataFrame, metric_cols: List[str], alpha: float = 0.05) -> pd.DataFrame:
    iterations = sorted(df["iteration"].unique())
    pairs = [(iterations[i], iterations[i + 1]) for i in range(len(iterations) - 1)]
    n_comparisons = len(pairs)
    corrected_alpha = alpha / max(n_comparisons, 1)
    rows = []
    for iter_a, iter_b in pairs:
        d_a = df[df["iteration"] == iter_a].sort_values("seed")
        d_b = df[df["iteration"] == iter_b].sort_values("seed")
        if len(d_a) != len(d_b) or len(d_a) < 2:
            continue
        for metric in metric_cols:
            v_a, v_b = d_a[metric].values, d_b[metric].values
            valid = ~(np.isnan(v_a) | np.isnan(v_b))
            if valid.sum() < 2:
                continue
            t_stat, p_value = scipy_stats.ttest_rel(v_a[valid], v_b[valid])
            diff = v_b[valid] - v_a[valid]
            d_cohen = float(np.mean(diff) / max(np.std(diff, ddof=1), EPS))
            rows.append({
                "iteration_from": iter_a, "iteration_to": iter_b, "metric": metric,
                "t_statistic": float(t_stat), "p_value": float(p_value),
                "p_value_bonferroni": float(min(p_value * n_comparisons, 1.0)),
                "significant_raw": p_value < alpha, "significant_bonferroni": p_value < corrected_alpha,
                "cohens_d": d_cohen, "mean_improvement": float(np.mean(diff)),
            })
    return pd.DataFrame(rows)


def compute_bootstrap_ci(values: np.ndarray, n_bootstrap: int = 1000, ci: float = 0.95, seed: int = 42) -> Dict[str, float]:
    rng = np.random.RandomState(seed)
    values = np.asarray(values, dtype=np.float64)
    values = values[~np.isnan(values)]
    if len(values) == 0:
        return {"mean": np.nan, "ci_low": np.nan, "ci_high": np.nan, "ci_width": np.nan}
    boot_means = np.array([np.mean(rng.choice(values, size=len(values), replace=True)) for _ in range(n_bootstrap)])
    alpha_half = (1 - ci) / 2
    ci_low = float(np.percentile(boot_means, alpha_half * 100))
    ci_high = float(np.percentile(boot_means, (1 - alpha_half) * 100))
    return {"mean": float(np.mean(values)), "ci_low": ci_low, "ci_high": ci_high, "ci_width": ci_high - ci_low}


def compute_all_bootstrap_cis(df: pd.DataFrame, group_cols: List[str], metric_cols: List[str], n_bootstrap: int = 1000) -> pd.DataFrame:
    rows = []
    for name, group in df.groupby(group_cols):
        if not isinstance(name, tuple):
            name = (name,)
        base = dict(zip(group_cols, name))
        for metric in metric_cols:
            vals = group[metric].dropna().values
            ci = compute_bootstrap_ci(vals, n_bootstrap=n_bootstrap)
            rows.append({
                **base, "metric": metric, "bootstrap_mean": ci["mean"],
                "bootstrap_ci_low": ci["ci_low"], "bootstrap_ci_high": ci["ci_high"],
                "bootstrap_ci_width": ci["ci_width"]
            })
    return pd.DataFrame(rows)


def make_calibration_comparison_table(results_df: pd.DataFrame, reference_iteration: int, reference_ratio: float) -> pd.DataFrame:
    subset = results_df[(results_df["iteration"] == reference_iteration) & (np.isclose(results_df["labeled_ratio"], reference_ratio))]
    return subset.groupby(["dataset", "calibration_method"], as_index=False)[
        ["NLL", "Brier", "ECE", "TACE", "SCE", "ACE", "MCE"]
    ].mean().sort_values(["dataset", "calibration_method"])


def build_multi_dataset_table(aggregated_df: pd.DataFrame, final_iteration: int, method_name: str) -> pd.DataFrame:
    subset = aggregated_df[(aggregated_df["iteration"] == final_iteration) & (aggregated_df["calibration_method"] == method_name)].copy()
    keep_cols = [
        "dataset", "labeled_ratio", "accuracy_mean_std", "ECE_mean_std", "ACE_mean_std",
        "SCE_mean_std", "TACE_mean_std", "NLL_mean_std", "Brier_mean_std",
        "entropy_mean_mean_std", "AURC_mean_std",
    ]
    missing = [col for col in keep_cols if col not in subset.columns]
    if missing:
        raise KeyError(f"Missing columns for multi-dataset table: {missing}")
    return subset[keep_cols].sort_values(["dataset", "labeled_ratio"])


def build_pvalue_matrix(sig_df: pd.DataFrame, metric: str, value_col: str = "p_value_bonferroni") -> pd.DataFrame:
    sub = sig_df[sig_df["metric"] == metric]
    methods = sorted(set(sub["method_1"].tolist() + sub["method_2"].tolist()))
    matrix = pd.DataFrame(np.ones((len(methods), len(methods))), index=methods, columns=methods)
    for _, row in sub.iterrows():
        matrix.loc[row["method_1"], row["method_2"]] = row[value_col]
        matrix.loc[row["method_2"], row["method_1"]] = row[value_col]
    np.fill_diagonal(matrix.values, 1.0)
    return matrix


# Fő futási blokk

if RUN_EXPERIMENTS:
    results_df, pseudo_df, gaps_df, reference_artifacts, cal_cv_df = run_all_experiments(ACTIVE_CONFIG)

    if len(results_df) == 0:
        raise RuntimeError("No experiment results were produced.")

    dataset_name = ACTIVE_CONFIG["datasets"][0]
    dataset_dirs = build_dataset_output_dirs(ACTIVE_CONFIG["output_dir"], dataset_name)
    summaries_dir = dataset_dirs["summaries"]
    plots_dir = summaries_dir / "plots"
    arrays_dir = summaries_dir / "arrays"
    stats_dir = summaries_dir / "statistics"
    plots_dir.mkdir(parents=True, exist_ok=True)
    arrays_dir.mkdir(parents=True, exist_ok=True)
    stats_dir.mkdir(parents=True, exist_ok=True)

    # Nyers és kötelező táblák mentése
    results_df.to_csv(summaries_dir / "results_raw.csv", index=False)
    pseudo_df.to_csv(summaries_dir / "pseudo_stats.csv", index=False)
    gaps_df.to_csv(summaries_dir / "generalization_gaps.csv", index=False)
    cal_cv_df.to_csv(summaries_dir / "calibration_cv_gaps.csv", index=False)
    results_df.to_csv(summaries_dir / "summary_results.csv", index=False)

elif RUN_ANALYSIS:
    # Ha nincs új futás, az elemzés a korábban mentett eredményeket tölti be.
    print("Csak elemzés fut, eredmények betöltése lemezről.")
    dataset_name = ACTIVE_CONFIG["datasets"][0]
    dataset_dirs = build_dataset_output_dirs(ACTIVE_CONFIG["output_dir"], dataset_name)
    summaries_dir = dataset_dirs["summaries"]
    plots_dir = summaries_dir / "plots"
    arrays_dir = summaries_dir / "arrays"
    stats_dir = summaries_dir / "statistics"
    plots_dir.mkdir(parents=True, exist_ok=True)
    arrays_dir.mkdir(parents=True, exist_ok=True)
    stats_dir.mkdir(parents=True, exist_ok=True)

    _raw_path = summaries_dir / "results_raw.csv"
    _pseudo_path = summaries_dir / "pseudo_stats.csv"
    _gaps_path = summaries_dir / "generalization_gaps.csv"
    _cal_cv_path = summaries_dir / "calibration_cv_gaps.csv"

    _all_datasets = ACTIVE_CONFIG["datasets"]
    _per_dataset_dfs = []
    _per_dataset_pseudo_dfs = []
    _per_dataset_gap_dfs = []
    for _ds_name in _all_datasets:
        _ds_dirs = build_dataset_output_dirs(ACTIVE_CONFIG["output_dir"], _ds_name)
        _ds_summary = _ds_dirs["summaries"] / "summary_results.csv"
        _ds_pseudo = _ds_dirs["summaries"] / "summary_pseudo_stats.csv"
        _ds_gaps = _ds_dirs["summaries"] / "summary_generalization_gaps.csv"
        if _ds_summary.exists():
            try:
                _per_dataset_dfs.append(pd.read_csv(_ds_summary))
                print(f"  Betöltve: {_ds_summary}")
            except Exception as _e:
                print(f"  Sérült CSV: {_ds_summary}. Hiba: {_e}")
        if _ds_pseudo.exists():
            try:
                _per_dataset_pseudo_dfs.append(pd.read_csv(_ds_pseudo))
            except Exception:
                pass
        if _ds_gaps.exists():
            try:
                _per_dataset_gap_dfs.append(pd.read_csv(_ds_gaps))
            except Exception:
                pass

    if _per_dataset_dfs:
        results_df = pd.concat(_per_dataset_dfs, ignore_index=True)
        pseudo_df = pd.concat(_per_dataset_pseudo_dfs, ignore_index=True) if _per_dataset_pseudo_dfs else pd.DataFrame()
        gaps_df = pd.concat(_per_dataset_gap_dfs, ignore_index=True) if _per_dataset_gap_dfs else pd.DataFrame()
        cal_cv_df = pd.read_csv(_cal_cv_path) if _cal_cv_path.exists() else pd.DataFrame()
    elif _raw_path.exists():
        try:
            results_df = pd.read_csv(_raw_path)
            pseudo_df = pd.read_csv(_pseudo_path) if _pseudo_path.exists() else pd.DataFrame()
            gaps_df = pd.read_csv(_gaps_path) if _gaps_path.exists() else pd.DataFrame()
            cal_cv_df = pd.read_csv(_cal_cv_path) if _cal_cv_path.exists() else pd.DataFrame()
        except Exception as exc:
            raise RuntimeError(
                f"Sérült CSV fájl a betöltéskor: {exc}. "
                f"Töröld a sérült fájlt, és futtasd újra RUN_EXPERIMENTS=True-val."
            ) from exc
    else:
        raise FileNotFoundError(
            "Nincs mentett eredmény egyik dataset mappájában sem. "
            "Futtasd előbb RUN_EXPERIMENTS=True-val."
        )
    reference_artifacts = {}
    print(f"  Betöltve: {len(results_df)} result sor, {len(pseudo_df)} pseudo sor")

else:
    print("RUN_EXPERIMENTS és RUN_ANALYSIS is False. Nincs teendő.")

# Elemzés

if RUN_EXPERIMENTS or RUN_ANALYSIS:

    required_results_df = results_df[[c for c in REQUIRED_RESULT_COLUMNS if c in results_df.columns]].copy()
    required_results_df.to_csv(summaries_dir / "results_required_columns.csv", index=False)

    # Aggregálás
    metric_cols = [col for col in REQUIRED_RESULT_COLUMNS if col not in {"dataset", "labeled_ratio", "iteration", "calibration_method", "seed"}]
    metric_cols = [c for c in metric_cols if c in required_results_df.columns]
    group_cols = ["dataset", "labeled_ratio", "iteration", "calibration_method"]
    aggregated_df = aggregate_mean_std(required_results_df, group_cols=group_cols, metric_cols=metric_cols)
    aggregated_df.to_csv(summaries_dir / "results_aggregated_mean_std.csv", index=False)

    formatted_cols = group_cols + [f"{col}_mean_std" for col in metric_cols]
    formatted_df = aggregated_df[[c for c in formatted_cols if c in aggregated_df.columns]].copy()
    formatted_df.to_csv(summaries_dir / "results_aggregated_formatted.csv", index=False)

    ci_cols = group_cols + [f"{col}_mean_ci" for col in metric_cols]
    ci_df = aggregated_df[[c for c in ci_cols if c in aggregated_df.columns]].copy()
    ci_df.to_csv(summaries_dir / "results_with_ci.csv", index=False)

    final_iteration = int(required_results_df["iteration"].max())
    reference_ratio = float(ACTIVE_CONFIG["labeled_ratios"][len(ACTIVE_CONFIG["labeled_ratios"]) // 2])
    pseudo_method = ACTIVE_CONFIG["pseudo_label_calibration_method"]

    calibration_table = make_calibration_comparison_table(required_results_df, final_iteration, reference_ratio)
    calibration_table.to_csv(summaries_dir / "calibration_comparison_table.csv", index=False)

    dataset_summary = aggregated_df.sort_values(["dataset", "labeled_ratio", "iteration", "calibration_method"]).reset_index(drop=True)
    dataset_summary.to_csv(summaries_dir / "dataset_summary.csv", index=False)

    ratio_table = aggregated_df[(aggregated_df["iteration"] == final_iteration) & (aggregated_df["calibration_method"] == pseudo_method)].copy()
    ratio_table.to_csv(summaries_dir / "ratio_effect_table.csv", index=False)

    round_table = aggregated_df[(np.isclose(aggregated_df["labeled_ratio"], reference_ratio)) & (aggregated_df["calibration_method"] == pseudo_method)].copy()
    round_table.to_csv(summaries_dir / "iteration_effect_table.csv", index=False)

    np.save(arrays_dir / "final_results_required.npy", required_results_df.to_records(index=False))
    np.save(arrays_dir / "final_results_aggregated.npy", aggregated_df.to_records(index=False))

    # Szignifikanciavizsgálat
    sig_alpha = ACTIVE_CONFIG["significance_alpha"]
    sig_subset = required_results_df[(required_results_df["iteration"] == final_iteration) & (np.isclose(required_results_df["labeled_ratio"], reference_ratio))]
    sig_metrics = ["accuracy", "ECE", "NLL", "Brier", "AURC", "MCC", "balanced_accuracy", "macro_F1"]
    sig_metrics = [m for m in sig_metrics if m in sig_subset.columns]

    if len(sig_subset) > 0:
        pairwise_sig = compute_pairwise_significance(sig_subset, "calibration_method", sig_metrics, alpha=sig_alpha)
        pairwise_sig.to_csv(stats_dir / "pairwise_calibration_significance.csv", index=False)
        print("\nKalibrációs módszerek páronkénti összehasonlítása Bonferroni korrekcióval:")
        if len(pairwise_sig) > 0:
            print(pairwise_sig[pairwise_sig["significant_bonferroni"]][["method_1", "method_2", "metric", "p_value_bonferroni", "cohens_d", "mean_diff"]].to_string(index=False))

        friedman_results = compute_friedman_test(sig_subset, "calibration_method", sig_metrics)
        friedman_results.to_csv(stats_dir / "friedman_calibration_test.csv", index=False)
        print("\nFriedman-teszt:")
        print(friedman_results.to_string(index=False))

    iter_sig_subset = required_results_df[(np.isclose(required_results_df["labeled_ratio"], reference_ratio)) & (required_results_df["calibration_method"] == pseudo_method)]
    if len(iter_sig_subset) > 0:
        iter_sig = compute_iteration_significance(iter_sig_subset, sig_metrics, alpha=sig_alpha)
        iter_sig.to_csv(stats_dir / "iteration_significance.csv", index=False)
        print("\nIterációk közti szignifikanciavizsgálat:")
        if len(iter_sig) > 0:
            print(iter_sig[["iteration_from", "iteration_to", "metric", "p_value", "cohens_d", "mean_improvement", "significant_raw"]].to_string(index=False))

    bootstrap_cis = compute_all_bootstrap_cis(required_results_df, group_cols=group_cols, metric_cols=sig_metrics, n_bootstrap=ACTIVE_CONFIG["n_bootstrap"])
    bootstrap_cis.to_csv(stats_dir / "bootstrap_confidence_intervals.csv", index=False)

    # Végső összefoglaló táblák
    final_summary = required_results_df.sort_values(["dataset", "labeled_ratio", "iteration", "calibration_method", "seed"]).reset_index(drop=True)
    final_summary.to_csv(summaries_dir / "final_statistical_summary.csv", index=False)
    formatted_df.to_csv(summaries_dir / "final_statistical_summary_mean_std.csv", index=False)
    ci_df.to_csv(summaries_dir / "final_statistical_summary_ci.csv", index=False)

    # H1: iterációk hatása
    print("\nH1: pseudo-label zajszint iterációnként")
    if len(pseudo_df) > 0 and "pseudo_noise_rate" in pseudo_df.columns:
        h1_agg = pseudo_df.groupby(["dataset", "labeled_ratio", "iteration"], as_index=False).agg(
            noise_mean=("pseudo_noise_rate", "mean"),
            noise_std=("pseudo_noise_rate", "std"),
            acceptance_mean=("acceptance_ratio", "mean"),
            accepted_count_mean=("accepted_count", "mean"),
        )
        h1_agg.to_csv(stats_dir / "h1_pseudo_noise_by_iteration.csv", index=False)

        plt.figure(figsize=(8, 5))
        for (ds, ratio), grp in h1_agg.groupby(["dataset", "labeled_ratio"]):
            grp = grp.sort_values("iteration")
            plt.errorbar(grp["iteration"], grp["noise_mean"], yerr=grp["noise_std"],
                         marker="o", capsize=4, label=f"{ds} r={ratio}")
        plt.xlabel("Iteráció")
        plt.ylabel("Pseudo-label zajszint")
        plt.title("H1: Pseudo-label zajszint iterációnként")
        plt.legend(fontsize=7)
        plt.grid(alpha=0.3)
        save_and_show_plot(plots_dir / "h1_noise_vs_iteration.png")

        plt.figure(figsize=(8, 5))
        for (ds, ratio), grp in h1_agg.groupby(["dataset", "labeled_ratio"]):
            grp = grp.sort_values("iteration")
            plt.errorbar(grp["iteration"], grp["acceptance_mean"],
                         marker="s", capsize=4, label=f"{ds} r={ratio}")
        plt.xlabel("Iteráció")
        plt.ylabel("Elfogadási arány")
        plt.title("H1: Pseudo-label elfogadási arány iterációnként")
        plt.legend(fontsize=7)
        plt.grid(alpha=0.3)
        save_and_show_plot(plots_dir / "h1_acceptance_vs_iteration.png")

        h1_acc = required_results_df[required_results_df["calibration_method"] == pseudo_method].copy()
        h1_acc_agg = h1_acc.groupby(["dataset", "labeled_ratio", "iteration"], as_index=False).agg(
            acc_mean=("accuracy", "mean"), acc_std=("accuracy", "std"),
        )
        plt.figure(figsize=(8, 5))
        for (ds, ratio), grp in h1_acc_agg.groupby(["dataset", "labeled_ratio"]):
            grp = grp.sort_values("iteration")
            plt.errorbar(grp["iteration"], grp["acc_mean"], yerr=grp["acc_std"],
                         marker="o", capsize=4, label=f"{ds} r={ratio}")
        plt.xlabel("Iteráció")
        plt.ylabel("Teszt pontosság")
        plt.title("H1: Teszt pontosság iterációnként")
        plt.legend(fontsize=7)
        plt.grid(alpha=0.3)
        save_and_show_plot(plots_dir / "h1_accuracy_vs_iteration.png")

    # tanár-diák javulás a címkézett arány függvényében
    print("teacher-student javulás elemzése")
    h2_method = pseudo_method
    h2_df = required_results_df[required_results_df["calibration_method"] == h2_method].copy()

    if len(h2_df) > 0:
        teacher_df = h2_df[h2_df["iteration"] == 0][["dataset", "labeled_ratio", "seed", "accuracy"]].rename(columns={"accuracy": "teacher_accuracy"})
        student_df = h2_df[h2_df["iteration"] == final_iteration][["dataset", "labeled_ratio", "seed", "accuracy"]].rename(columns={"accuracy": "student_accuracy"})
        gain_df = teacher_df.merge(student_df, on=["dataset", "labeled_ratio", "seed"])
        gain_df["gain"] = gain_df["student_accuracy"] - gain_df["teacher_accuracy"]
        gain_df.to_csv(stats_dir / "h2_teacher_student_gain.csv", index=False)

        gain_agg = gain_df.groupby(["dataset", "labeled_ratio"], as_index=False).agg(
            gain_mean=("gain", "mean"), gain_std=("gain", "std"),
            teacher_mean=("teacher_accuracy", "mean"), student_mean=("student_accuracy", "mean"),
        )
        gain_agg.to_csv(stats_dir / "h2_gain_aggregated.csv", index=False)
        print(gain_agg.to_string(index=False))

        plt.figure(figsize=(8, 5))
        for ds, grp in gain_agg.groupby("dataset"):
            grp = grp.sort_values("labeled_ratio")
            plt.errorbar(grp["labeled_ratio"], grp["gain_mean"], yerr=grp["gain_std"],
                         marker="o", capsize=4, label=ds, linewidth=1.5)
        plt.axhline(y=0, color="gray", linestyle="--", alpha=0.5)
        plt.xlabel("Jelölt adatok aránya")
        plt.ylabel("Pontosságjavulás (diák – tanár)")
        plt.title("H2: Tanár–diák javulás a jelölt arány függvényében")
        plt.legend()
        plt.grid(alpha=0.3)
        save_and_show_plot(plots_dir / "h2_gain_vs_ratio.png")

        plt.figure(figsize=(8, 5))
        for ds, grp in gain_agg.groupby("dataset"):
            grp = grp.sort_values("labeled_ratio")
            plt.plot(grp["labeled_ratio"], grp["teacher_mean"], marker="s", linestyle="--", label=f"{ds} tanár")
            plt.plot(grp["labeled_ratio"], grp["student_mean"], marker="o", linestyle="-", label=f"{ds} diák")
        plt.xlabel("Jelölt adatok aránya")
        plt.ylabel("Teszt pontosság")
        plt.title("H2: Tanár vs diák pontosság")
        plt.legend(fontsize=7)
        plt.grid(alpha=0.3)
        save_and_show_plot(plots_dir / "h2_teacher_vs_student.png")

        for ds in gain_df["dataset"].unique():
            ds_gains = gain_df[gain_df["dataset"] == ds]
            ratio_groups = [grp["gain"].values for _, grp in ds_gains.groupby("labeled_ratio")]
            if len(ratio_groups) >= 3 and all(len(g) >= 2 for g in ratio_groups):
                try:
                    stat, pval = scipy_stats.kruskal(*ratio_groups)
                    print(f"  {ds}: Kruskal-Wallis (gain ~ ratio): H={stat:.3f}, p={pval:.4f}")
                except Exception:
                    pass

    # szűrési stratégiák összehasonlítása
    print("\nH3: szűrési stratégiák összehasonlítása")
    if len(pseudo_df) > 0 and "no_filter_noise_rate" in pseudo_df.columns:
        h3_records = []
        for _, row in pseudo_df.iterrows():
            base = {
                "dataset": row["dataset"], "labeled_ratio": row["labeled_ratio"],
                "iteration": row["iteration"], "seed": row["seed"]
            }
            h3_records.append({
                **base, "strategy": "no_filter",
                "noise_rate": row["no_filter_noise_rate"],
                "accepted_count": row["no_filter_accepted_count"],
                "acceptance_ratio": 1.0
            })
            h3_records.append({
                **base, "strategy": "confidence_only",
                "noise_rate": row["confidence_only_noise_rate"],
                "accepted_count": row["confidence_only_accepted_count"],
                "acceptance_ratio": row["confidence_only_acceptance_ratio"]
            })
            h3_records.append({
                **base, "strategy": "entropy_calibration",
                "noise_rate": row["pseudo_noise_rate"],
                "accepted_count": row["accepted_count"],
                "acceptance_ratio": row["acceptance_ratio"]
            })
        h3_df = pd.DataFrame(h3_records)
        h3_df.to_csv(stats_dir / "h3_filtering_strategies.csv", index=False)

        h3_agg = h3_df.groupby(["dataset", "labeled_ratio", "iteration", "strategy"], as_index=False).agg(
            noise_mean=("noise_rate", "mean"), noise_std=("noise_rate", "std"),
            accept_ratio_mean=("acceptance_ratio", "mean"),
        )
        h3_agg.to_csv(stats_dir / "h3_filtering_aggregated.csv", index=False)

        strategies = ["no_filter", "confidence_only", "entropy_calibration"]
        strategy_labels = ["Szűrés nélkül", "Csak konfidencia", "Entrópia + kalibráció"]
        colors = ["#e74c3c", "#f39c12", "#27ae60"]
        for ds in h3_agg["dataset"].unique():
            ds_data = h3_agg[(h3_agg["dataset"] == ds) & (h3_agg["iteration"] == 0)]
            if len(ds_data) == 0:
                continue
            ratios = sorted(ds_data["labeled_ratio"].unique())
            x = np.arange(len(ratios))
            w = 0.25
            fig, ax = plt.subplots(figsize=(9, 5))
            for i, (strat, label, color) in enumerate(zip(strategies, strategy_labels, colors)):
                vals = ds_data[ds_data["strategy"] == strat].set_index("labeled_ratio").reindex(ratios)
                ax.bar(x + i * w, vals["noise_mean"].values, w, yerr=vals["noise_std"].values,
                       label=label, color=color, capsize=3)
            ax.set_xticks(x + w)
            ax.set_xticklabels([f"{r:.0%}" for r in ratios])
            ax.set_xlabel("Jelölt adatok aránya")
            ax.set_ylabel("Pseudo-label zajszint")
            ax.set_title(f"Szűrési stratégiák összehasonlítása ({ds})")
            ax.legend()
            ax.grid(alpha=0.3, axis="y")
            save_and_show_plot(plots_dir / f"h3_filtering_noise_{ds}.png")

        plt.figure(figsize=(8, 5))
        markers = {"no_filter": "X", "confidence_only": "s", "entropy_calibration": "o"}
        for strat, label, color in zip(strategies, strategy_labels, colors):
            sub = h3_agg[h3_agg["strategy"] == strat]
            plt.scatter(sub["accept_ratio_mean"], sub["noise_mean"],
                        marker=markers[strat], color=color, s=60, label=label, alpha=0.8)
        plt.xlabel("Elfogadási arány")
        plt.ylabel("Pseudo-label zajszint")
        plt.title("Szűrési trade-off (elfogadás vs zaj)")
        plt.legend()
        plt.grid(alpha=0.3)
        save_and_show_plot(plots_dir / "h3_filtering_tradeoff.png")

        for ds in h3_df["dataset"].unique():
            ds_h3 = h3_df[h3_df["dataset"] == ds]
            strat_pairs = [("no_filter", "entropy_calibration"),
                           ("confidence_only", "entropy_calibration"),
                           ("no_filter", "confidence_only")]
            for s1, s2 in strat_pairs:
                v1 = ds_h3[ds_h3["strategy"] == s1].sort_values(["labeled_ratio", "seed"])["noise_rate"].values
                v2 = ds_h3[ds_h3["strategy"] == s2].sort_values(["labeled_ratio", "seed"])["noise_rate"].values
                valid = ~(np.isnan(v1) | np.isnan(v2))
                if valid.sum() >= 2:
                    t_stat, p_val = scipy_stats.ttest_rel(v1[valid], v2[valid])
                    print(f"  {ds}: {s1} vs {s2}: t={t_stat:.3f}, p={p_val:.4f}, Δnoise={np.mean(v1[valid]-v2[valid]):.4f}")

    k1_methods = ["uncalibrated", "static_temperature", "temperature_scaling"]
    k1_metrics = ["ECE", "NLL", "Brier", "accuracy"]
    k1_df = required_results_df[required_results_df["calibration_method"].isin(k1_methods)].copy()

    if len(k1_df) > 0:
        k1_agg = k1_df.groupby(["dataset", "labeled_ratio", "iteration", "calibration_method"], as_index=False)[k1_metrics].mean()
        k1_table = k1_agg[k1_agg["iteration"] == final_iteration].copy()
        k1_table.to_csv(stats_dir / "k1_temperature_comparison.csv", index=False)
        print(k1_table.to_string(index=False))

        for ds in k1_table["dataset"].unique():
            ds_k1 = k1_table[k1_table["dataset"] == ds]
            ratios = sorted(ds_k1["labeled_ratio"].unique())
            x = np.arange(len(ratios))
            w = 0.25
            fig, ax = plt.subplots(figsize=(8, 5))
            method_labels = {
                "uncalibrated": "Kalibrálatlan (T=1)",
                "static_temperature": "Statikus (T=1.5)",
                "temperature_scaling": "Dinamikus (tanult T)"
            }
            method_colors = {
                "uncalibrated": "#e74c3c",
                "static_temperature": "#f39c12",
                "temperature_scaling": "#27ae60"
            }
            for i, m in enumerate(k1_methods):
                vals = ds_k1[ds_k1["calibration_method"] == m].set_index("labeled_ratio").reindex(ratios)
                if len(vals) > 0:
                    ax.bar(x + i * w, vals["ECE"].values, w, label=method_labels.get(m, m), color=method_colors.get(m, "gray"), capsize=3)
            ax.set_xticks(x + w)
            ax.set_xticklabels([f"{r:.0%}" for r in ratios])
            ax.set_xlabel("Jelölt adatok aránya")
            ax.set_ylabel("ECE")
            ax.set_title(f"Hőmérséklet-skálázás hatása az ECE-re ({ds})")
            ax.legend()
            ax.grid(alpha=0.3, axis="y")
            save_and_show_plot(plots_dir / f"k1_ece_comparison_{ds}.png")

        k1_sig_subset = k1_df[(k1_df["iteration"] == final_iteration)]
        for ds in k1_sig_subset["dataset"].unique():
            ds_sub = k1_sig_subset[k1_sig_subset["dataset"] == ds]
            static_vals = ds_sub[ds_sub["calibration_method"] == "static_temperature"].sort_values(["labeled_ratio", "seed"])
            dynamic_vals = ds_sub[ds_sub["calibration_method"] == "temperature_scaling"].sort_values(["labeled_ratio", "seed"])
            for metric in ["ECE", "NLL"]:
                v_s = static_vals[metric].values
                v_d = dynamic_vals[metric].values
                valid = ~(np.isnan(v_s) | np.isnan(v_d))
                if valid.sum() >= 2 and len(v_s) == len(v_d):
                    t_stat, p_val = scipy_stats.ttest_rel(v_s[valid], v_d[valid])
                    print(f"  {ds}: static vs dynamic ({metric}): t={t_stat:.3f}, p={p_val:.4f}, Δ={np.mean(v_s[valid]-v_d[valid]):.4f}")
    print("Kész.")
```

```python
# -*- coding: utf-8 -*-
"""
H3 hipotézis vizsgálat — Google Colab script.

Másolj be egy Colab cellába és futtasd!

H3: Az entrópia-alapú bizonytalansági szűrés és a kalibrált valószínűségek
    együttes alkalmazása azonos vagy összevethető lefedettség mellett
    megbízhatóbb pszeudocímke-kiválasztást eredményez, mint a kizárólag
    maximális valószínűségi küszöbön alapuló szelekció.

Vizsgálatok:
  1) Noise–Coverage görbék: 3 szelekciós stratégia (kalibrálatlan konf.,
     kalibrált konf., kalibrált konf.+entrópia) összehasonlítása
  2) Fair összehasonlítás: azonos lefedettségen a 3 stratégia zajrátája
  3) Entrópia-mechanizmus: a high-confidence poolon belül az entrópia
     rangsorolás extra információt ad-e a zajszűrésben
  4) Párosított statisztikai tesztek + hatásméret

Megjegyzés: argmax(softmax(logits/T)) == argmax(logits) bármely T > 0-ra,
ezért a predikciók (és hibák) MINDEN stratégiánál azonosak — csak a
KIVÁLASZTÁSI SZABÁLY különbözik.
"""


def h3_hipotezis_vizsgalat():
    import gc
    import json
    import re
    import numpy as np
    import pandas as pd
    import matplotlib.pyplot as plt
    import tensorflow as tf
    from pathlib import Path
    from scipy import stats as sp_stats
    from scipy.stats import rankdata
    from sklearn.model_selection import train_test_split
    from tensorflow.keras import layers
    from tensorflow.keras.applications import EfficientNetB0
    from tensorflow.keras.applications.efficientnet import preprocess_input
    from tensorflow.keras.models import Model
    from tensorflow.keras.optimizers import Adam
    from tensorflow.keras.losses import SparseCategoricalCrossentropy

    # ── Drive mount ──────────────────────────────────────────
    try:
        from google.colab import drive
        drive.mount("/content/drive", force_remount=False)
    except Exception:
        pass

    # ── Paraméterek ──────────────────────────────────────────
    BASE       = Path("/content/drive/MyDrive/tdk/tdk_final")
    DS_ALL     = ["cifar10", "svhn", "cifar100"]
    DS_LABELS  = {"cifar10": "CIFAR-10", "svhn": "SVHN", "cifar100": "CIFAR-100"}
    SEEDS      = [42, 1337, 2024]
    N_ITER     = 3
    VAL_RATIO  = 0.1
    MAX_TRAIN  = 30_000
    IMG, BS    = 224, 128
    CONF_TH    = 0.9
    CONF_INC   = 0.02
    ENT_PCT    = 30.0
    ENT_DEC    = 5.0
    EPS        = 1e-12
    CAL_METHOD = "temperature_scaling"
    N_PTS      = 25          # pontszám a görbékhez

    # ── Segédfüggvények ──────────────────────────────────────

    def _softmax(logits):
        logits = logits - np.max(logits, axis=1, keepdims=True)
        e = np.exp(logits)
        return np.clip(e / np.sum(e, axis=1, keepdims=True), EPS, 1.0)

    def _entropy(probs):
        p = np.clip(probs, EPS, 1.0)
        return -np.sum(p * np.log(p), axis=1)

    def _build_model(nc):
        bb = EfficientNetB0(
            weights="imagenet", include_top=False,
            input_shape=(IMG, IMG, 3),
        )
        bb.trainable = False
        f = layers.GlobalAveragePooling2D()(bb.output)
        f = layers.Dropout(0.2)(f)
        o = layers.Dense(nc, activation=None, dtype="float32")(f)
        m = Model(bb.input, o)
        m.compile(Adam(1e-4), SparseCategoricalCrossentropy(from_logits=True))
        return m

    def _make_ds(x):
        ds = tf.data.Dataset.from_tensor_slices(x)
        ds = ds.map(
            lambda img: preprocess_input(
                tf.image.resize(tf.cast(img, tf.float32), (IMG, IMG))),
            num_parallel_calls=tf.data.AUTOTUNE,
        )
        return ds.batch(BS).prefetch(tf.data.AUTOTUNE)

    def _subsample(x, y, n, seed):
        if n and len(y) > n:
            idx, _ = train_test_split(
                np.arange(len(y)), train_size=n, stratify=y, random_state=seed)
            return x[idx], y[idx]
        return x, y

    def _load_temp(ds_dir, run_name):
        p = ds_dir / "checkpoints" / f"{run_name}_{CAL_METHOD}.json"
        if p.exists():
            with open(p) as f:
                return json.load(f).get("temperature", 1.0)
        return 1.0

    def _load_dataset(name):
        if name == "cifar10":
            (x, y), _ = tf.keras.datasets.cifar10.load_data()
            return x.astype(np.uint8), y.flatten().astype(np.int32), 10
        if name == "cifar100":
            (x, y), _ = tf.keras.datasets.cifar100.load_data(label_mode="fine")
            return x.astype(np.uint8), y.flatten().astype(np.int32), 100
        if name == "svhn":
            try:
                import tensorflow_datasets as tfds
                imgs, lbls = [], []
                for im, lb in tfds.as_numpy(
                        tfds.load("svhn_cropped", split="train",
                                  as_supervised=True)):
                    imgs.append(im)
                    lbls.append(int(lb))
                return (np.stack(imgs).astype(np.uint8),
                        np.array(lbls, np.int32), 10)
            except Exception as e:
                print(f"  SVHN betöltés sikertelen: {e}")
        return None, None, None

    def _noise_at_topk(errors, scores, n_points):
        """Noise-Coverage görbe: magasabb score = először kiválasztva."""
        order = np.argsort(-scores)
        sorted_errs = errors[order]
        n = len(errors)
        covs, noises = [], []
        for frac in np.linspace(0.05, 1.0, n_points):
            k = max(1, int(round(frac * n)))
            covs.append(k / n)
            noises.append(float(np.mean(sorted_errs[:k])))
        return np.array(covs), np.array(noises)

    # ═════════════════════════════════════════════════════════
    # ADATGYŰJTÉS  (modell-újratöltéssel)
    # ═════════════════════════════════════════════════════════
    fair_records = []          # per (ds, ratio, seed, iter) fair comparison
    curve_data   = {}          # ds → [per-seed noise-coverage curves]
    mech_data    = {}          # ds → [per-seed entropy-mechanism curves]

    for ds_name in DS_ALL:
        ds_dir = BASE / ds_name
        if not ds_dir.exists():
            print(f"[SKIP] {DS_LABELS.get(ds_name, ds_name)}: "
                  f"nem található ({ds_dir})")
            continue

        # Ratio-k felderítése
        wt_files = list(ds_dir.glob("*_final.weights.h5"))
        if not wt_files:
            wt_files = list((ds_dir / "checkpoints").glob("*_best.weights.h5"))
        ratios_pct = sorted({
            int(m.group(1))
            for f in wt_files
            for m in [re.search(r"ratio(\d+)", f.name)] if m
        })
        if not ratios_pct:
            print(f"[SKIP] {DS_LABELS.get(ds_name, ds_name)}: nincs súlyfájl")
            continue

        mid_r = ratios_pct[len(ratios_pct) // 2]

        x_tr, y_tr, nc = _load_dataset(ds_name)
        if x_tr is None:
            continue

        tf.keras.backend.clear_session()
        model = _build_model(nc)
        curve_data[ds_name] = []
        mech_data[ds_name] = []

        print(f"\n{'=' * 55}")
        print(f"  {DS_LABELS[ds_name]}  ({nc} osztály, ratio-k: {ratios_pct})")
        print(f"{'=' * 55}")

        for ratio_pct in ratios_pct:
            ratio_frac = ratio_pct / 100.0
            for seed in SEEDS:
                # ── Adatfelosztás reprodukálása ──────────────
                x_s, y_s = _subsample(x_tr, y_tr, MAX_TRAIN, seed)
                tr_idx, _ = train_test_split(
                    np.arange(len(y_s)), test_size=VAL_RATIO,
                    stratify=y_s, random_state=seed)
                x_pool, y_pool = x_s[tr_idx], y_s[tr_idx]
                _, unl_idx = train_test_split(
                    np.arange(len(y_pool)), train_size=ratio_frac,
                    stratify=y_pool, random_state=seed)
                x_unl = x_pool[unl_idx]
                y_unl = y_pool[unl_idx]

                remaining = np.arange(len(y_unl))

                for it in range(N_ITER):
                    run = f"{ds_name}_ratio{ratio_pct}_seed{seed}_iter{it}"

                    # Súlyfájl keresése
                    wt = ds_dir / f"{run}_final.weights.h5"
                    if not wt.exists():
                        wt = ds_dir / "checkpoints" / f"{run}_best.weights.h5"
                    if not wt.exists():
                        print(f"    [SKIP] {run}: nincs súly")
                        break

                    model.load_weights(str(wt))
                    x_rem = x_unl[remaining]
                    y_rem = y_unl[remaining]

                    logits = model.predict(
                        _make_ds(x_rem), verbose=0
                    ).astype(np.float32)
                    temp = _load_temp(ds_dir, run)

                    # Predikciók (argmax invariáns T-re)
                    pred = np.argmax(logits, axis=1)
                    errors = (pred != y_rem).astype(np.float32)

                    # Kalibrálatlan valószínűségek (T = 1)
                    probs_uncal = _softmax(logits)
                    conf_uncal = np.max(probs_uncal, axis=1)

                    # Kalibrált valószínűségek (tanult T)
                    probs_cal = _softmax(logits / max(temp, 1e-6))
                    conf_cal = np.max(probs_cal, axis=1)
                    ent_cal = _entropy(probs_cal)

                    # ── Pipeline kettős szűrése ──────────────
                    eff_c = min(CONF_TH + it * CONF_INC, 0.99)
                    eff_p = max(ENT_PCT - it * ENT_DEC, 5.0)
                    ent_th = float(np.percentile(ent_cal, eff_p))

                    dual_mask = (conf_cal >= eff_c) & (ent_cal <= ent_th)
                    dual_count = int(np.sum(dual_mask))
                    dual_noise = (float(np.mean(errors[dual_mask]))
                                  if dual_count > 0 else np.nan)
                    dual_cov = (dual_count / len(y_rem)
                                if len(y_rem) > 0 else 0.0)

                    # ── Fair összehasonlítás: azonos coverage ─
                    if dual_count > 0:
                        top_cal = np.argsort(-conf_cal)[:dual_count]
                        top_uncal = np.argsort(-conf_uncal)[:dual_count]
                        noise_cal_only = float(np.mean(errors[top_cal]))
                        noise_uncal_only = float(np.mean(errors[top_uncal]))
                    else:
                        noise_cal_only = noise_uncal_only = np.nan

                    fair_records.append({
                        "dataset": ds_name,
                        "ratio_pct": ratio_pct,
                        "iteration": it,
                        "seed": seed,
                        "coverage": dual_cov,
                        "n_selected": dual_count,
                        "n_pool": len(y_rem),
                        "noise_dual": dual_noise,
                        "noise_cal_only": noise_cal_only,
                        "noise_uncal_only": noise_uncal_only,
                        "temperature": temp,
                    })

                    # ── Görbék (csak középső ratio, iter 0) ──
                    if ratio_pct == mid_r and it == 0:
                        # Noise-Coverage görbék
                        cov_pts, n_uncal = _noise_at_topk(
                            errors, conf_uncal, N_PTS)
                        _, n_cal = _noise_at_topk(
                            errors, conf_cal, N_PTS)
                        # Kombinált score: rank(conf) + rank(-entropy)
                        combined = (rankdata(conf_cal) +
                                    rankdata(-ent_cal))
                        _, n_dual_c = _noise_at_topk(
                            errors, combined, N_PTS)
                        curve_data[ds_name].append({
                            "cov": cov_pts,
                            "n_uncal": n_uncal,
                            "n_cal": n_cal,
                            "n_dual": n_dual_c,
                        })

                        # Entrópia-mechanizmus (high-conf poolon belül)
                        hc_idx = np.where(conf_cal >= eff_c)[0]
                        if len(hc_idx) >= 20:
                            hc_ent = ent_cal[hc_idx]
                            hc_err = errors[hc_idx]
                            order_low = np.argsort(hc_ent)
                            order_high = np.argsort(-hc_ent)
                            overall_hc = float(np.mean(hc_err))
                            fracs = np.linspace(0.05, 1.0, N_PTS)
                            nl, nh = [], []
                            for frac in fracs:
                                k = max(1, int(round(
                                    frac * len(hc_idx))))
                                nl.append(float(np.mean(
                                    hc_err[order_low[:k]])))
                                nh.append(float(np.mean(
                                    hc_err[order_high[:k]])))
                            mech_data[ds_name].append({
                                "fracs": fracs * 100,
                                "noise_low": np.array(nl),
                                "noise_high": np.array(nh),
                                "overall": overall_hc,
                                "hc_count": len(hc_idx),
                            })

                    print(f"    {run}:  dual={dual_noise:.4f}  "
                          f"cal_conf={noise_cal_only:.4f}  "
                          f"uncal={noise_uncal_only:.4f}  "
                          f"cov={dual_cov:.3f}  T={temp:.3f}")

                    # Pool szűkítése
                    remaining = np.delete(
                        remaining, np.where(dual_mask)[0])

        del model
        gc.collect()

    if not fair_records:
        print("\nNincs megjeleníthető adat!")
        return

    df = pd.DataFrame(fair_records)
    avail = [d for d in DS_ALL if d in df["dataset"].unique()]
    n = len(avail)
    plt.rcParams.update({"font.size": 11})

    # ═════════════════════════════════════════════════════════
    #  1. ÁBRA: Noise–Coverage görbék
    # ═════════════════════════════════════════════════════════
    fig1, axes1 = plt.subplots(1, n, figsize=(6 * n, 5), squeeze=False)

    STRATS_CURVE = [
        ("n_uncal", "Kalibrálatlan konfidencia",        "#e74c3c", "--"),
        ("n_cal",   "Kalibrált konfidencia",            "#f39c12", "-"),
        ("n_dual",  "Kalibrált konf. + entrópia",       "#27ae60", "-"),
    ]

    for ci, ds in enumerate(avail):
        ax = axes1[0, ci]
        curves = curve_data.get(ds, [])
        if not curves:
            ax.set_title(f"{DS_LABELS[ds]}\n(nincs adat)")
            continue

        cov = curves[0]["cov"]
        for key, label, color, ls in STRATS_CURVE:
            arr = np.array([c[key] for c in curves])
            mu = arr.mean(axis=0)
            sigma = arr.std(axis=0)
            ax.plot(cov * 100, mu, label=label,
                    color=color, ls=ls, lw=2)
            ax.fill_between(cov * 100, mu - sigma, mu + sigma,
                            alpha=0.15, color=color)

        mid_r = sorted(df[df["dataset"] == ds]["ratio_pct"].unique())
        mid_r = mid_r[len(mid_r) // 2]
        ax.set_title(f"{DS_LABELS[ds]}  (jelölt: {mid_r}%, iter 0)",
                     fontsize=12, fontweight="bold")
        ax.set_xlabel("Lefedettség (%)")
        if ci == 0:
            ax.set_ylabel("Pszeudocímke zajráta")
        ax.grid(alpha=0.3)
        ax.legend(fontsize=8, loc="lower right")

    fig1.suptitle(
        "H3: Szelekciós stratégiák Noise–Coverage görbéi\n"
        "(alacsonyabb görbe = megbízhatóbb szelekció "
        "azonos lefedettségen)",
        fontsize=13, fontweight="bold",
    )
    fig1.tight_layout(rect=[0, 0, 1, 0.87])
    plt.show()

    # ═════════════════════════════════════════════════════════
    #  2. ÁBRA: Fair összehasonlítás + entrópia-mechanizmus
    # ═════════════════════════════════════════════════════════
    fig2, axes2 = plt.subplots(2, n, figsize=(5.5 * n, 9),
                               squeeze=False)

    for ci, ds in enumerate(avail):
        sub = df[df["dataset"] == ds]
        ratios = sorted(sub["ratio_pct"].unique())
        mid_r = ratios[len(ratios) // 2]
        sub_mid = sub[sub["ratio_pct"] == mid_r]
        iters = sorted(sub_mid["iteration"].unique())

        # ── Felső sor: bar chart (fair summary iterációnként) ─
        ax = axes2[0, ci]
        x = np.arange(len(iters))
        w = 0.25
        STRATS_BAR = [
            ("noise_uncal_only",
             "Kalibrálatlan\nkonfidencia",   "#e74c3c"),
            ("noise_cal_only",
             "Kalibrált\nkonfidencia",       "#f39c12"),
            ("noise_dual",
             "Kalibrált konf.\n+ entrópia",  "#27ae60"),
        ]
        for si, (col, label, color) in enumerate(STRATS_BAR):
            mu = sub_mid.groupby("iteration")[col] \
                        .mean().reindex(iters)
            sd = sub_mid.groupby("iteration")[col] \
                        .std().reindex(iters).fillna(0)
            bars = ax.bar(x + si * w, mu, w, yerr=sd,
                          label=label, color=color,
                          capsize=3, alpha=0.85,
                          edgecolor="black", linewidth=0.5)
            for b, v in zip(bars, mu):
                if not np.isnan(v):
                    ax.text(b.get_x() + b.get_width() / 2,
                            b.get_height() + 0.003,
                            f"{v:.3f}", ha="center",
                            va="bottom", fontsize=7)

        # Lefedettség az x-tick címkékbe
        mean_covs = [
            sub_mid[sub_mid["iteration"] == it]["coverage"].mean()
            for it in iters
        ]
        ax.set_xticks(x + w)
        ax.set_xticklabels(
            [f"{it + 1}\n(cov: {c:.0%})"
             for it, c in zip(iters, mean_covs)])
        ax.set_title(f"{DS_LABELS[ds]}  (jelölt: {mid_r}%)",
                     fontsize=12, fontweight="bold")
        ax.set_xlabel("Iteráció")
        ax.grid(alpha=0.3, axis="y")
        if ci == 0:
            ax.set_ylabel("Zajráta azonos lefedettségen\n"
                          "(alacsonyabb = jobb)")
        if ci == n - 1:
            ax.legend(fontsize=7, loc="upper right")

        # ── Alsó sor: entrópia-mechanizmus ───────────────────
        ax2 = axes2[1, ci]
        mechs = mech_data.get(ds, [])
        if mechs:
            fracs = mechs[0]["fracs"]
            nl_arr = np.array([m["noise_low"] for m in mechs])
            nh_arr = np.array([m["noise_high"] for m in mechs])
            ov_mean = np.mean([m["overall"] for m in mechs])
            hc_mean = int(np.mean([m["hc_count"] for m in mechs]))

            for arr, label, color in [
                (nl_arr,
                 "Alacsony entrópia először", "#27ae60"),
                (nh_arr,
                 "Magas entrópia először",    "#e74c3c"),
            ]:
                mu = arr.mean(axis=0)
                sd = arr.std(axis=0)
                ax2.plot(fracs, mu, label=label,
                         color=color, lw=2, marker="o", ms=3)
                ax2.fill_between(fracs, mu - sd, mu + sd,
                                 alpha=0.12, color=color)

            ax2.axhline(ov_mean, ls="--", lw=1.5, color="gray",
                        label=f"Pool átlaga ({ov_mean:.3f})")
            eff_c0 = min(CONF_TH, 0.99)
            ax2.set_title(
                f"Entrópia rangsorolás ereje "
                f"(conf ≥ {eff_c0:.0%}, "
                f"~{hc_mean} minta)",
                fontsize=10)
            ax2.set_xlabel("Elfogadott rész a "
                           "high-conf poolból (%)")
            if ci == 0:
                ax2.set_ylabel("Pszeudocímke zajráta")
            ax2.grid(alpha=0.3)
            ax2.legend(fontsize=8)
        else:
            ax2.set_title("(nincs adat)")

    fig2.suptitle(
        "H3: Fair összehasonlítás azonos lefedettségen  |  "
        "Entrópia-mechanizmus",
        fontsize=13, fontweight="bold")
    fig2.tight_layout(rect=[0, 0, 1, 0.94])
    plt.show()

    # ═════════════════════════════════════════════════════════
    #  STATISZTIKAI TESZTEK
    # ═════════════════════════════════════════════════════════
    print("\n" + "=" * 90)
    print("H3 HIPOTÉZIS — STATISZTIKAI VIZSGÁLAT")
    print("=" * 90)

    for ds in avail:
        sub = df[df["dataset"] == ds]
        print(f"\n{'─' * 70}")
        print(f"  {DS_LABELS[ds]}")
        print(f"{'─' * 70}")

        d = sub["noise_dual"].values
        c_cal = sub["noise_cal_only"].values
        c_uncal = sub["noise_uncal_only"].values

        # ── Test 1: Kalibrált konf. vs kettős szűrés (H3 mag) ─
        v = ~(np.isnan(d) | np.isnan(c_cal))
        if v.sum() >= 3:
            diff = c_cal[v] - d[v]
            t, p = sp_stats.ttest_rel(c_cal[v], d[v])
            cd = float(np.mean(diff) /
                       max(np.std(diff, ddof=1), EPS))
            print(f"\n  1) Kalibrált konfidencia vs Kettős szűrés "
                  f"(párosított t-próba, N={v.sum()}):")
            print(f"     Δzajráta  = {np.mean(diff):.5f} "
                  f"± {np.std(diff):.5f}")
            print(f"     t({v.sum()-1}) = {t:.3f},  "
                  f"p = {p:.6f},  Cohen's d = {cd:.3f}")
            sig1 = p < 0.05 and np.mean(diff) > 0
            print(f"     → {'✓ H3 ALÁTÁMASZTVA' if sig1 else '✗ H3 nem alátámasztva'} (α = 0.05)")

        # ── Test 2: Kalibrálatlan konf. vs kettős szűrés ──────
        v2 = ~(np.isnan(d) | np.isnan(c_uncal))
        if v2.sum() >= 3:
            diff2 = c_uncal[v2] - d[v2]
            t2, p2 = sp_stats.ttest_rel(c_uncal[v2], d[v2])
            cd2 = float(np.mean(diff2) /
                        max(np.std(diff2, ddof=1), EPS))
            print(f"\n  2) Kalibrálatlan konfidencia vs Kettős szűrés "
                  f"(N={v2.sum()}):")
            print(f"     Δzajráta  = {np.mean(diff2):.5f} "
                  f"± {np.std(diff2):.5f}")
            print(f"     t({v2.sum()-1}) = {t2:.3f},  "
                  f"p = {p2:.6f},  Cohen's d = {cd2:.3f}")

        # ── Dekompozíció: kalibráció vs entrópia hozzájárulás ─
        v3 = ~(np.isnan(c_uncal) | np.isnan(c_cal))
        v4 = ~(np.isnan(c_cal) | np.isnan(d))
        if v3.sum() >= 3 and v4.sum() >= 3:
            d_cal = c_uncal[v3] - c_cal[v3]
            d_ent = c_cal[v4] - d[v4]
            print(f"\n  Dekompozíció:")
            print(f"     Kalibráció hozzájárulása:  "
                  f"Δ = {np.mean(d_cal):.5f} "
                  f"({np.mean(d_cal)/(max(np.mean(c_uncal[v3]),EPS))*100:+.1f}%)")
            print(f"     Entrópia hozzájárulása:    "
                  f"Δ = {np.mean(d_ent):.5f} "
                  f"({np.mean(d_ent)/(max(np.mean(c_cal[v4]),EPS))*100:+.1f}%)")

        # ── Per-iteration test (kalibrált konf. vs dual) ──────
        for it_val in sorted(sub["iteration"].unique()):
            si = sub[sub["iteration"] == it_val]
            dd = si["noise_dual"].values
            cc = si["noise_cal_only"].values
            vi = ~(np.isnan(dd) | np.isnan(cc))
            if vi.sum() >= 3:
                di = cc[vi] - dd[vi]
                ti, pi = sp_stats.ttest_rel(cc[vi], dd[vi])
                print(f"\n  Iter {it_val + 1}:  Δ = {np.mean(di):.5f},  "
                      f"p = {pi:.4f},  "
                      f"{'✓' if pi < 0.05 and np.mean(di) > 0 else '✗'}")

    # ═════════════════════════════════════════════════════════
    #  ÖSSZEFOGLALÓ TÁBLÁZAT
    # ═════════════════════════════════════════════════════════
    print("\n\n" + "=" * 100)
    print("ÖSSZEFOGLALÓ TÁBLÁZAT  "
          "(seed-ek átlaga ± szórás, azonos lefedettségen)")
    print("=" * 100)

    for ds in avail:
        sub = df[df["dataset"] == ds]
        for ratio in sorted(sub["ratio_pct"].unique()):
            sr = sub[sub["ratio_pct"] == ratio]
            print(f"\n{DS_LABELS[ds]}  –  jelölt: {ratio}%")
            hdr = (f"{'It':>3} | {'Lefedettség':>11} | "
                   f"{'Kal.-lan konf.':>16} | "
                   f"{'Kalibrált konf.':>16} | "
                   f"{'Kettős szűrés':>16} | "
                   f"{'Zajcsökk.':>10}")
            print(hdr)
            print("─" * len(hdr))
            for it_v in sorted(sr["iteration"].unique()):
                r = sr[sr["iteration"] == it_v]
                cov_str = f"{r['coverage'].mean():.1%}"

                def _f(col, rr=r):
                    m, s = rr[col].mean(), rr[col].std()
                    if np.isnan(m):
                        return "       n/a      "
                    return f"{m:.4f}±{s:.4f}"

                mc = r["noise_cal_only"].mean()
                md = r["noise_dual"].mean()
                red = (((mc - md) / max(mc, EPS)) * 100
                       if not (np.isnan(mc) or np.isnan(md))
                       else np.nan)
                red_str = (f"{red:+.1f}%"
                           if not np.isnan(red) else "  n/a")
                print(f"{it_v + 1:>3} | {cov_str:>11} | "
                      f"{_f('noise_uncal_only'):>16} | "
                      f"{_f('noise_cal_only'):>16} | "
                      f"{_f('noise_dual'):>16} | "
                      f"{red_str:>10}")

    print("\n✔ H3 hipotézis vizsgálat kész.")


# ── Futtatás ──────────────────────────────────────────────────
h3_hipotezis_vizsgalat()
```
